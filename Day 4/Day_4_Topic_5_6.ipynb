```python
#Day 4, Topic 5: The .transform() Method
import pandas as pd

df = pd.read_csv('titanic.csv')
```

```python
# .transform() – returns same number of rows as original DataFrame
df['AvgAge_Pclass'] = df.groupby('Pclass')['Age'].transform('mean')
df[['Pclass', 'Age', 'AvgAge_Pclass']].head(10)
```

```python
#Common Use Case – Filling Missing Values with Group Mean
# Fill missing Age with the median Age of that passenger's Pclass
df['Age_filled'] = df.groupby('Pclass')['Age'].transform(
    lambda x: x.fillna(x.median())
)

df[['Pclass', 'Age', 'Age_filled']].head(10)
```

```python
#Group‑Wise Standardization (Z‑Score)
df['Fare_zscore'] = df.groupby('Pclass')['Fare'].transform(
    lambda x: (x - x.mean()) / x.std()
)

df[['Pclass', 'Fare', 'Fare_zscore']].head(10)
```

```python
#Custom Functions with .transform()
df['Age_rank'] = df.groupby('Pclass')['Age'].transform(
    lambda x: x.rank(ascending=False)
)

df[['Pclass', 'Age', 'Age_rank']].head(10)
```

```python
#Practice Activity: .transform()

import pandas as pd
df = pd.read_csv('titanic.csv')
```

```python
"""Task 1: Use .transform() to create a column AvgFare_Sex that contains the average fare for each passenger's gender. 
    Display Sex, Fare, and AvgFare_Sex for the first 5 rows."""

df['AvgFare_Sex'] = df.groupby('Sex')['Fare'].transform('mean')

df[['Sex', 'Fare', 'AvgFare_Sex']].head()
```

```python
"""Task 2: Create a column Fare_diff that is the difference between a passenger's Fare and 
the median fare of their Pclass. Use .transform() with 'median'."""

df['Fare_diff'] = df['Fare'] - df.groupby('Pclass')['Fare'].transform('median')

df[['Pclass', 'Fare', 'Fare_diff']]
```

```python
"""Task 3: Fill missing Age values with the mean age of the passenger's Pclass and Sex group. 
Use .transform() with a custom lambda that fills NaN."""

df['Age'] = df.groupby(['Pclass', 'Sex'])['Age'].transform(lambda x: x.fillna(x.mean()))

# Verify there are no more missing values in Age
print(f"Missing Age values: {df['Age'].isnull().sum()}")
```

```python
"""Task 4: Create a column Fare_rank that ranks fares within each Embarked port (descending, so highest fare = rank 1). 
Use .transform().
"""

df['Fare_Rank'] = df.groupby('Embarked')['Fare'].transform(lambda x: x.rank(ascending=False, method='min'))
df[['Embarked', 'Fare', 'Fare_Rank']].sort_values(['Embarked', 'Fare_Rank']).head(10)
```

```python
"""Task 5 (Bonus): Create a column Age_zscore that standardizes Age within each Pclass (subtract group mean, 
divide by group std). Then filter the DataFrame to show only rows where Age_zscore is greater than 2 (outliers)."""

df['Age_zscore'] = df.groupby('Pclass')['Age'].transform(lambda x: (x - x.mean()) / x.std())

outliers = df[df['Age_zscore'] > 2]

print(len(outliers))
outliers[['Pclass', 'Name', 'Age', 'Age_zscore']].head()
```

```python
#Day 4, Topic 6: The .filter() Method for Groups
import pandas as pd
df = pd.read_csv('titanic.csv')
```

```python
#Basic Syntax
filtered = df.groupby('Pclass').filter(lambda g: len(g) >= 200)
print(len(df))
print(len(filtered))
```

```python
#Filtering Smaller Groups

embarked_counts = df.groupby('Embarked').size()
print(embarked_counts)

filtered = df.groupby('Embarked').filter(lambda g: len(g) >= 100)
len(filtered)
```

```python
#Filtering Based on Group Statistics
#Example: Keep only classes where the average fare is greater than 30.


filtered = df.groupby('Pclass').filter(lambda g: g['Fare'].mean() > 30)
filtered['Pclass'].value_counts()
```

```python
#Using a Named Function for Clarity
def group_has_survivors(g):
    return g['Survived'].sum() > 0

filtered = df.groupby('Sex').filter(group_has_survivors)
filtered['Sex'].value_counts()
```

```python
#Practice Activity: .filter() on Groups 
import pandas as pd
df = pd.read_csv('titanic.csv')
```

```python
"""Task 1: Group by Embarked and use .filter() to keep only ports that have at least 100 passengers.
How many rows remain? Verify that 'Q' is no longer in the Embarked column."""

filtered = df.groupby('Embarked').filter(lambda g: len(g) >= 100)

print(len(filtered))
filtered['Embarked'].unique()
```

```python
"""Task 2: Group by Pclass and keep only classes where the average age is greater than 30. Which class(es) remain? 
Display the value counts of Pclass in the filtered DataFrame."""

filtered = df.groupby('Pclass').filter(lambda g: g['Age'].mean() > 30)

print(filtered['Pclass'].unique())

print(filtered['Pclass'].value_counts())
```

```python
"""Task 3: Group by Sex and Pclass (together). Use .filter() to keep only subgroups that have more than 
50 passengers. How many rows remain?"""

filtered = df.groupby(['Sex', 'Pclass']).filter(lambda g: len(g) > 50)

len(filtered)
```

```python
"""Task 4: Create a new column FamilySize = SibSp + Parch + 1. Group by FamilySize and keep 
only family sizes that have at least 10 passengers. What are the remaining family sizes?"""

df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
filtered = df.groupby('FamilySize').filter(lambda g: len(g) >= 10)

print(filtered['FamilySize'].unique())
len(filtered)
```

```python
"""Task 5 (Bonus): Group by Embarked and keep only ports where the survival rate is greater than 35%. 
Which ports remain? Show the survival rate for each remaining port."""

df_high_survival = df.groupby('Embarked').filter(lambda x: x['Survived'].mean() > 0.35)

remaining_rates = df_high_survival.groupby('Embarked')['Survived'].mean()

print("Ports with a survival rate > 35%:")
print(remaining_rates)
```
