```python
#⚔️ Gauntlet #4: Titanic Grouping Mastery
import pandas as pd
import numpy as np

df = pd.read_csv('titanic.csv')
```

```python
# Task 1: Group by Pclass with named aggregations
task1 = df.groupby('Pclass').agg(
    avg_age=('Age', 'mean'),
    total_fare=('Fare', 'sum'),
    survivors=('Survived', 'sum')
)
print("Task 1 - Named Aggregations by Pclass:\n", task1)
```

```python
# Task 2: Group by Sex and Embarked, use dict for different aggregations
task2 = df.groupby(['Sex', 'Embarked']).agg({
    'Age': ['mean', 'median'],
    'Fare': 'sum',
    'Survived': 'mean'
})
print("\nTask 2 - Dict agg by Sex and Embarked:\n", task2)
```

```python
# Task 3: Fill missing Age with median of Pclass and Sex group
df['Age'] = df.groupby(['Pclass', 'Sex'])['Age'].transform(
    lambda x: x.fillna(x.median())
)
print("\nTask 3 - Missing Age filled. Remaining nulls:", df['Age'].isna().sum())
```

```python
# Task 4: Keep Embarked ports with ≥200 passengers AND survival rate > 35%
filtered = df.groupby('Embarked').filter(
    lambda g: (len(g) >= 200) and (g['Survived'].mean() > 0.35)
)
remaining_ports = filtered['Embarked'].unique()
print("\nTask 4 - Ports meeting conditions:", remaining_ports)
```

```python
# Task 5: Pivot table - average Fare by Pclass and Embarked, fill NaN with 0
pivot = pd.pivot_table(
    df,
    values='Fare',
    index='Pclass',
    columns='Embarked',
    aggfunc='mean',
    fill_value=0
)
print("\nTask 5 - Pivot Table (Avg Fare):\n", pivot)
```

```python
# Task 6: Crosstab Pclass vs Survived, row proportions, with margins
crosstab = pd.crosstab(
    index=df['Pclass'],
    columns=df['Survived'],
    normalize='index',
    margins=True,
    margins_name='Total'
)
print("\nTask 6 - Crosstab (row proportions):\n", crosstab)
```

```python
# Task 7: IQR of Fare per Pclass using a custom function
def iqr(x):
    return x.quantile(0.75) - x.quantile(0.25)

iqr_result = df.groupby('Pclass')['Fare'].agg(iqr).rename('Fare_IQR')
print("\nTask 7 - Fare IQR by Pclass:\n", iqr_result)
```

```python
# Task 8 (Bonus): For each Pclass, percentage of passengers by Embarked
pct_by_class = df.groupby('Pclass')['Embarked'].value_counts(normalize=True).mul(100).round(2)
pct_by_class = pct_by_class.rename('percentage').reset_index()
print("\nTask 8 (Bonus) - Embarked % within each Pclass:\n", pct_by_class)
```
