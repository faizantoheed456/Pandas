```python
#Day 4, Topic 3: Aggregating Multiple Columns

import pandas as pd
df = pd.read_csv('titanic.csv')
```

```python
#Dictionary Approach – One Function per Column
result = df.groupby('Pclass').agg({
    'Age': 'mean', 
    'Fare': 'sum',
    'Survived': 'sum'
})

result
```

```python
#Multiple Functions per Column – List Approach
result = df.groupby('Pclass').agg({
    'Age': ['mean', 'median', 'std'],
    'Fare': ['sum', 'max']
})

result
```

```python
#Mixed Aggregations – Single and Multiple
result = df.groupby('Embarked').agg({
    'Age': 'median',
    'Fare': ['mean', 'max'],
    'Survived': 'sum'
})

result
```

```python
#Named Aggregations (Clean Column Names)
result = df.groupby('Sex').agg(
    avg_age=('Age', 'mean'),
    total_fare=('Fare', 'sum'),
    survivors=('Survived', 'sum'),
    survival_pct=('Survived', 'mean')
)

result
```

```python
#Grouping by Multiple Columns with Named Aggregations
result = df.groupby(['Pclass', 'Sex']).agg(
    count=('PassengerId', 'count'),
    avg_age=('Age', 'mean'),
    avg_fare=('Fare', 'mean'),
    survived=('Survived', 'sum')
)

result
```

```python
#Practice Activity: Multi‑Column Aggregations
import pandas as pd
df = pd.read_csv('titanic.csv')
```

```python
"""Task 1: Group by Pclass and compute:
Average Age
Total Fare
Number of survivors (Survived sum)
Use a dictionary with single functions for each column."""

pclass_stats = df.groupby('Pclass').agg({
    'Age': 'mean',
    'Fare': 'sum',
    'Survived': 'sum'
})

pclass_stats
```

```python
"""Task 2: Group by Embarked. For Age, compute mean and min. For Fare, compute sum and max. 
Use a dictionary with lists for multiple functions."""

embarked_stats = df.groupby('Embarked').agg({
    'Age': ['mean', 'min'],
    'Fare': ['sum', 'max'],
})

embarked_stats
```

```python
"""Task 3: Use named aggregations to group by Sex and produce a DataFrame with columns:
passenger_count (count of rows)
avg_fare
max_age
survival_rate (mean of Survived)"""

result = df.groupby('Sex').agg(
    passenger_count=('PassengerId', 'count'),
    avg_fare=('Fare', 'mean'),
    max_age=('Age', 'max'),
    survival_rate=('Survived', 'mean')
)

result
```

```python
"""Task 4: Group by both Pclass and Embarked. Use named aggregations to compute:
total_passengers (count)
avg_fare
total_survivors (sum of Survived)"""

result = df.groupby(['Pclass', 'Embarked']).agg(
    total_passengers=('PassengerId', 'count'),
    avg_fare=('Fare', 'mean'),
    total_survivors=('Survived', 'sum')
)

result
```

```python
"""Task 5 (Bonus): Create a new column AgeGroup with bins [0, 18, 35, 60, 100] and labels 
['Child', 'Young Adult', 'Adult', 'Senior']. Then group by AgeGroup and Sex, and compute the average 
fare and survival rate using named aggregations."""

df['AgeGroup'] = pd.cut(df['Age'], bins=[0, 18, 35, 60, 100], labels=['Child', 'Young Adult', 'Adult', 'Senior'])

result = df.groupby(['AgeGroup', 'Sex']).agg(
    avg_fare=('Fare', 'mean'),
    survival_rate=('Survived', 'mean')
)

result
```

```python
#Day 4, Topic 4: Grouping by Multiple Columns
import pandas as pd

df = pd.read_csv('titanic.csv')
```

```python
#Basic Multi‑Column GroupBy
grouped = df.groupby(['Pclass', 'Sex'])
avg_fare = grouped['Fare'].mean()
avg_fare
```

```python
#Accessing Results from Multi‑Column GroupBy
avg_fare.loc[(1, 'female')]
```

```python
# All stats for first class (both sexes)
avg_fare.loc[1]
```

```python
#Flattening the Result with .reset_index()
avg_fare_flat = avg_fare.reset_index(name='avg_fare')
avg_fare_flat
```

```python
#Using .agg() with Multiple Columns
result = df.groupby(['Pclass', 'Sex']).agg(
    total_passengers=('PassengerId', 'count'),
    avg_age=('Age', 'mean'),
    avg_fare=('Fare', 'mean'),
    survival_rate=('Survived', 'mean')
).reset_index()

result
```

```python
#Grouping by Three or More Columns
result = df.groupby(['Pclass', 'Sex', 'Embarked']).agg(
    count=('PassengerId', 'count'),
    avg_fare=('Fare', 'mean')
).reset_index()

result
```

```python
#The as_index=False Parameter
result = df.groupby(['Pclass', 'Sex'], as_index=False)['Fare'].mean()
result
```

```python
#Practice Activity: Multi‑Column GroupBy

import pandas as pd
df = pd.read_csv('titanic.csv')
```

```python
#Task 1: Group by Pclass and Embarked. Compute the average fare for each subgroup. Use .reset_index() to 
# flatten the result.

result = df.groupby(['Pclass', 'Embarked'])['Fare'].mean().reset_index()
result
```

```python
"""Task 2: Group by Sex and Pclass. Use .agg() to compute:
Number of passengers (count)
Average Age
Survival rate (mean of Survived)
Flatten the result with .reset_index()."""

result = df.groupby(['Sex', 'Pclass']).agg(
    number_of_passengers=('PassengerId', 'count'),
    avg_age=('Age', 'mean'),
    survival_rate=('Survived', 'mean')
).reset_index()

result
```

```python
"""Task 3: Use as_index=False to group by Embarked and Sex, and compute the total fare collected per subgroup. 
Which subgroup paid the most total fare?"""

result = df.groupby(['Embarked', 'Sex'], as_index=False)['Fare'].sum()
result

top_sub = result.loc[result['Fare'].idxmax()]

top_sub
```

```python
"""Task 4: Create a new column FamilySize = SibSp + Parch + 1. Then group by three columns: 
Pclass, Sex, and FamilySize. Compute the count of passengers in each subgroup. 
Sort by count descending. Which subgroup is most common?"""

df['FamilySize'] = df['SibSp'] + df['Parch'] + 1

result = df.groupby(['Pclass', 'Sex', 'FamilySize'])['PassengerId'].count().sort_values(ascending=False)
result.index[0]
```

```python
"""Task 5 (Bonus): Group by Pclass and Embarked. Use a dictionary inside .agg() to compute:
Age: ['mean', 'min', 'max']
Fare: 'sum'
Survived: 'mean'
Flatten the result and rename columns appropriately."""

result = df.groupby(['Pclass', 'Embarked']).agg(
    {
        'Age':['mean', 'min', 'max'],
        'Fare': 'sum',
        'Survived': 'mean'
    }
).reset_index()
result.columns = ['_'.join(col).strip() for col in result.columns.values]
result.rename(columns={
    'Age': 'Age_stats',
    'Fare': 'Total_Fare',
    'Survived': 'Survival_Rate'
})
```
