```python
#Day 4, Topic 9: Cross‑Tabulation (pd.crosstab)

import pandas as pd

df = pd.read_csv('titanic.csv')
```

```python
#Basic pd.crosstab()
crosstab = pd.crosstab(index=df['Pclass'], columns=df['Sex'])
crosstab
```

```python
#Adding Margins (Totals)
crosstab = pd.crosstab(index=df['Pclass'], columns=df['Sex'], margins=True, margins_name='Total')
crosstab
```

```python
#Normalization
crosstab_row_pct = pd.crosstab(df['Pclass'], df['Sex'], normalize='index')
crosstab_row_pct
```

```python
#Column percentages: What % of females are in each class?

crosstab_row_pct = pd.crosstab(df['Pclass'], df['Sex'], normalize='columns')
crosstab_row_pct
```

```python
#Multiple Index/Columns
crosstab = pd.crosstab(index=[df['Pclass'], df['Survived']], columns=df['Embarked'], margins=True)
crosstab
```

```python
#Values Other Than Counts
crosstab = pd.crosstab(index=df['Pclass'], columns=df['Sex'], values=df['Fare'], aggfunc='mean')

crosstab
```

```python
#Practice Activity: pd.crosstab
import pandas as pd

df = pd.read_csv('titanic.csv')
```

```python
#Task 1: Use pd.crosstab() to create a frequency table of Survived (rows) vs. Pclass (columns). Add margins.

crosstab = pd.crosstab(index=df['Survived'], columns=df['Pclass'], margins=True)
crosstab
```

```python
#Task 2: Create the same table but with row percentages (normalize='index'). 
# What percentage of survivors were in first class?

crosstab = pd.crosstab(index=df['Survived'], columns=df['Pclass'], normalize='index')

crosstab
crosstab.loc[1, 1] * 100
```

```python
#Task 3: Create a multi‑index crosstab with index=['Pclass', 'Sex'] and columns='Survived'. Display the result.
crosstab = pd.crosstab(index=[df['Pclass'], df['Sex']], columns=df['Survived'])

crosstab
```

```python
"""Task 4: Use pd.crosstab() to compute the median Age for each combination of Pclass (rows) 
and Sex (columns). (Hint: pass values='Age' and aggfunc='median'.)
"""

crosstab = pd.crosstab(
    index=df['Pclass'],
    columns=df['Sex'],
    values=df['Age'],
    aggfunc='median'
)

crosstab
```

```python
"""Task 5 (Bonus): Create a crosstab of Embarked (rows) and Survived (columns), normalized by columns, 
to show survival rate by port. Which port had the highest survival rate?
"""

crosstab = pd.crosstab(index=df['Embarked'], columns=df['Survived'], normalize='columns')
crosstab[1].idxmax()
```
