# 📊 Day 1, Topic 2 — The Series Object: Deep Dive

A `pd.Series` is the fundamental 1-D data structure in Pandas. It is essentially a **NumPy array with an index**.

```
Values:  [100, 200, 300]
Index:   [ 'a', 'b', 'c' ]
```

Every Series has:
- `.values` — the underlying NumPy array
- `.index` — the axis labels (default `RangeIndex(0, N)`)
- `.dtype` — the data type of the values
- `.name` — optional label for the Series itself


````python
#Day 1, Topic 2: The Series Object – Deep Dive
import pandas as pd
````

## Creating a Series

There are three main ways to construct a Series:

| Method | When to use |
|--------|-------------|
| From a list | Simple numeric or string data |
| From a dict | When keys should become index labels |
| From a scalar | Initialize all positions to the same value |


````python
#From a list
s1 = pd.Series([10, 20, 30])

print(s1)
````

````python
#From a dictionary (keys become index):

s2 = pd.Series({'a': 100, 'b': 200, 'c': 300})
print(s2)
````

````python
#With a scalar value (repeats for each index)

s3 = pd.Series(5, index=['a', 'b', 'c'])
print(s3)
````

## Accessing Elements: `.loc` vs `.iloc`

This is one of the most important distinctions in Pandas:

| Accessor | Basis | Endpoint inclusive? |
|----------|-------|---------------------|
| `.loc[label]` | **Label** (index name) | Yes — `s.loc['a':'c']` includes `'c'` |
| `.iloc[pos]` | **Position** (integer 0, 1, 2…) | No — `s.iloc[1:4]` excludes position 4 |

> **Rule:** Use `.loc` when you know the label; use `.iloc` when you know the position.  
> Mixing them up is the #1 source of bugs for Pandas beginners.


````python
#Accessing Elements
#By label (.loc):
print(s2.loc['b'])
````

````python
#By position (.iloc):
print(s2.iloc[1])
````

## Slicing — Label vs Position

Label slicing with `s['b':'d']` is **inclusive on both ends** (unlike Python lists).  
Position slicing with `s.iloc[1:4]` is **exclusive on the right** (like Python lists).

This asymmetry catches many beginners off guard — always double-check which you're using.


````python
#Slicing by label (inclusive of both ends!)
s4 = pd.Series([1, 2, 3, 4, 5], index=['a', 'b', 'c', 'd', 'e'])
print(s4['b':'d'])
````

````python
#Slicing by position (exclusive of end, like Python lists):
print(s4.iloc[1:4])
````

## Vectorized Operations

Series supports the full suite of NumPy-style vectorized operations:

```python
s * 2      # multiply every element
s + s      # add corresponding elements
s ** 2     # square every element
s > 2      # boolean mask — True/False per element
```

All of these run without Python loops — they execute in compiled C code via NumPy.


````python
#Vectorized Operations on Series
s = pd.Series([1, 2, 3, 4])
print(s * 2)

print(s + s)

print(s ** 2)

print(s > 2)
````

## Essential Series Methods

These are the methods you'll use daily when exploring a new Series:

| Method | What it does |
|--------|-------------|
| `.fillna(value)` | Replace `NaN` with a given value |
| `.head(n)` | First n elements |
| `.tail(n)` | Last n elements |
| `.describe()` | Count, mean, std, min, quartiles, max |
| `.value_counts()` | Frequency of each unique value |
| `.sort_values()` | Sort by value (ascending by default) |
| `.sort_index()` | Sort by index label |
| `.isna()` | Boolean mask: True where NaN |

> **NaN (Not a Number):** Pandas uses `NaN` (a special float) to represent missing data. Most Pandas operations skip `NaN` automatically — e.g., `.mean()` ignores it.


````python
s5 = pd.Series([1, 2, 3, 1, None, 1, 7, 8, 9])

print(s5.fillna(0))

print(s5.head(3))

print(s5.tail(5))

print(s5.describe())

print(s5.value_counts())

print(s5.sort_values(ascending=False))

print(s5.sort_index())

print(s5.isna())
````

## 📝 Practice Tasks

### Task 1 — Series from a Dict (Fruit Prices)
Dictionary keys become the index labels. This is the most readable way to create a labeled Series.


````python
# Create a Series from this dictionary: {'Apple': 3.5, 'Banana': 2.0, 'Cherry': 5.0}
# Print the Series.
fruit_stocks = pd.Series({'Apple': 3.5, 'Banana': 2.0, 'Cherry': 5.0})

print(fruit_stocks)
````

### Task 2 — `.loc`, `.iloc`, and Slicing
Practice the four access patterns on a single Series. Notice that `.loc['q':'s']` gives 3 elements (inclusive), while `.iloc[1:3]` gives only 2 (exclusive end).


````python
# Using the Series: s = pd.Series([100, 200, 300, 400], index=['p', 'q', 'r', 's'])
# - Access the element with label 'r' using .loc
# - Access the element at position 2 using .iloc
# - Slice from label 'q' to 's' (inclusive)
# - Slice from position 1 to 3 (exclusive end)

s = pd.Series([100, 200, 300, 400], index=['p', 'q', 'r', 's'])

print(s.loc['r'])
print(s.iloc[2])
print(s['q':'s'])
print(s.iloc[1:3])
````

### Task 3 — Boolean Masking
`temps > 35` creates a boolean Series.  
Using it inside `temps[...]` filters to only the rows where it's `True`.  
This two-step pattern — **create mask → apply mask** — is the foundation of all Pandas filtering.


````python
# Given: temps = pd.Series([32, 35, 28, 42, 38])
# Create a Boolean Series that is True where temperature > 35.
# Use that Boolean Series to filter the original Series (show only temps > 35).
temps = pd.Series([32, 35, 28, 42, 38])

print(temps > 35)

print(temps[temps > 35])
````

### Task 4 — Count and Fill Missing Values
`s.isna().sum()` counts missing values by summing the boolean `True`s.  
`s.mean()` automatically ignores `NaN`, so it's safe to use as the fill value.  
`s.fillna(mean)` returns a new Series — the original `s` is unchanged.


````python
# Create a Series: s = pd.Series([10, None, 30, None, 50])
# - Count how many missing values it has.
# - Fill missing values with the mean of the non‑missing values.
# Print the filled Series.
s = pd.Series([10, None, 30, None, 50])
print(s.isna().sum())
mean = s.mean()
print(s.fillna(mean))
````

### Task 5 — Sort, Describe, Find Max
`.sort_values(ascending=False)` ranks from highest to lowest.  
`.describe()` gives the "5-number summary" plus count and standard deviation.  
`.idxmax()` returns the **index label** of the maximum value — more useful than just the max number.


````python
# Given: grades = pd.Series([88, 92, 79, 85, 95], index=['Ali', 'Sara', 'Omar', 'Lina', 'Zain'])
# - Sort the Series by values (descending).
# - Print summary statistics using .describe().
# - Who has the highest grade? (Print the index label of the max value.)
grades = pd.Series([88, 92, 79, 85, 95], index=['Ali', 'Sara', 'Omar', 'Lina', 'Zain'])

print(grades.sort_values(ascending=False))
print(grades.describe())
print(grades.idxmax())
````
