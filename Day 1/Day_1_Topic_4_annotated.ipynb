# 📊 Day 1, Topic 4 — Inspecting Data: Essential Methods

When you receive any new dataset, your first job is **not** to analyze it — it's to *understand* it. These inspection methods are your toolkit for that first-contact phase.

## The Inspection Checklist
1. `df.head()` / `df.tail()` — see what rows look like
2. `df.shape` — how big is it?
3. `df.columns` / `df.index` — what are the labels?
4. `df.dtypes` — are types correct (e.g., numbers stored as strings)?
5. `df.info()` — non-null counts → find missing values immediately
6. `df.describe()` — numeric summary → spot outliers, zero-range columns
7. `df['col'].value_counts()` — category distributions
8. `df.nunique()` — how many unique values per column?

These 8 calls give you an 80% understanding of any dataset in under a minute.


````python
#Day 1, Topic 4: Inspecting Data – Essential Methods
import pandas as pd
````

````python
import pandas as pd

data = {
    'Name': ['Faizan', 'Ali', 'Usman', 'Arslan', 'Rohan'],
    'Age': [25, 30, 35, 28, None],
    'Salary': [50000, 60000, 70000, 55000, 65000],
    'Department': ['HR', 'IT', 'IT', 'HR', 'Finance']
}
df = pd.DataFrame(data)
````

````python
df
````

## `.head()` and `.tail()` — Peek at the Data

```python
df.head(n)   # first n rows (default: 5)
df.tail(n)   # last n rows (default: 5)
```

Always look at **both ends** — the start and end of a file can have very different data, headers, or formatting artifacts.


````python
print(df.head(2))

print(df.tail(2))
````

## `.shape`, `.columns`, `.index` — Structure at a Glance

```python
df.shape    # → (5, 4) — tuple: (num_rows, num_cols)
df.columns  # → Index(['Name', 'Age', 'Salary', 'Department'])
df.index    # → RangeIndex(start=0, stop=5, step=1)
```

`df.shape[0]` gives row count, `df.shape[1]` gives column count — useful in code.


````python
print(df.shape)
print(df.columns)
print(df.index)
````

## `.dtypes` and `.info()` — Types and Missingness

### `.dtypes`
Shows the dtype of every column. Common dtypes:
| Pandas dtype | Meaning |
|-------------|---------|
| `int64` | 64-bit integer |
| `float64` | 64-bit float (also used when column has NaN!) |
| `object` | strings or mixed types |
| `bool` | True/False |
| `datetime64[ns]` | dates/times |

### `.info()`
Prints a concise summary including **Non-Null Count** — the fastest way to spot missing data.  
`Age: 4 non-null float64` immediately tells you 1 row is missing an age.

> **Why float64 for Age?** NumPy integers can't represent `NaN`. When a column has any missing value, Pandas automatically upcasts integers to float64.


````python
print(df.dtypes)

print(df.info())
````

## `.describe()` — Statistical Summary

`.describe()` runs on **numeric columns** by default:

| Stat | Meaning |
|------|---------|
| `count` | Non-null count (not the same as total rows if there are NaNs!) |
| `mean` | Arithmetic mean |
| `std` | Standard deviation |
| `min` / `max` | Range |
| `25%`, `50%`, `75%` | Quartiles (50% = median) |

Use it to quickly spot: negative values that should be positive, salary of 0, ages of 999, etc.


````python
print(df.describe())
````

## `.describe(include='object')` — Categorical Summary

Pass `include='object'` to describe string/categorical columns:

| Stat | Meaning |
|------|---------|
| `count` | Non-null count |
| `unique` | Number of distinct values |
| `top` | Most frequent value |
| `freq` | How many times `top` appears |


````python
print(df.describe(include='object'))
````

## `.value_counts()` — Frequency Distribution

`.value_counts()` on a column gives you the count of each unique value, sorted from most to least frequent.  
Equivalent to SQL's `SELECT col, COUNT(*) GROUP BY col ORDER BY COUNT(*) DESC`.


````python
print(df['Department'].value_counts())
````

## `.nunique()` — Count of Unique Values per Column

`df.nunique()` returns a Series where each entry is how many distinct values that column has.  
- High nunique on a string column → likely a free-text field (not a category)
- Low nunique → likely categorical, good candidate for `pd.Categorical` dtype
- `nunique() == len(df)` → column is a unique ID


````python
print(df.nunique())
````

## 📝 Practical Activity — Employee Dataset


````python
#Practical Activity: Topic 4
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'Age': [25, 30, 35, 28, None],
    'Salary': [50000, 60000, 70000, 55000, 65000],
    'Department': ['HR', 'IT', 'IT', 'HR', 'Finance']
}
df = pd.DataFrame(data)
````

````python
df
````

### Task 1 — Quick Overview
`.tail(3)` shows the last 3 rows. Combined with `.shape`, you now know the dataset's full extent.


````python
"""Task 1: Quick Overview
Print the last 3 rows.
Print the shape."""

print(df.tail(3))
print(df.shape)
````

### Task 2 — Column Information
`.columns.tolist()` converts the `Index` object to a plain Python list — easier to read and copy.  
`.dtypes` tells you which columns might need type conversion (e.g., `Age` is `float64` because of the `NaN`).


````python
"""Task 2: Column Information
Print the column names.
Print the data types.
"""

print(df.columns.tolist())
print(df.dtypes)
````

### Task 3 — Missing Value Check
`.info()` is the fastest way to find missing data — look for columns where `Non-Null Count < total rows`.  
Here `Age: 4 non-null` tells us 1 value is missing out of 5 rows.


````python
"""Task 3: Missing Value Check
Use .info() to identify which column has missing values and how many.
"""
print(df.info())
````

### Task 4 — Numeric Summary + Median
`.describe()` is your first sanity check. The median (50th percentile) is more robust than the mean when outliers are present — it won't be skewed by one unusually high salary.


````python
"""Task 4: Numeric Summary
Print summary statistics for numeric columns.
What is the median age?"""
print(df.describe())
print(df['Age'].median())
````

### Task 5 — Categorical Summary
`.nunique()` counts distinct departments.  
`.value_counts().idxmax()` chains two calls: `.value_counts()` counts each department, `.idxmax()` returns the label of the maximum count.


````python
"""Task 5: Categorical Summary
How many unique departments are there?
Which department appears most frequently, and how many times?
"""

print(df['Department'].nunique())
top_dept = df['Department'].value_counts().idxmax()
top_freq = df['Department'].value_counts().max()
print(f"Most frequent department: '{top_dept}' appears {top_freq} times.")
````

### Task 6 — Value Distribution
`.value_counts()` on `'Department'` shows exactly how many employees are in each department — essential before any group-level analysis.


````python
"""Task 6: Value Counts
Use .value_counts() on the 'Department' column to see the distribution."""

print(df['Department'].value_counts())
````
