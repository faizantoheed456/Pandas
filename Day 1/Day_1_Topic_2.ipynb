```python
#Day 1, Topic 2: The Series Object – Deep Dive
import pandas as pd
```

```python
#From a list
s1 = pd.Series([10, 20, 30])

print(s1)
```

```python
#From a dictionary (keys become index):

s2 = pd.Series({'a': 100, 'b': 200, 'c': 300})
print(s2)
```

```python
#With a scalar value (repeats for each index)

s3 = pd.Series(5, index=['a', 'b', 'c'])
print(s3)
```

```python
#Accessing Elements
#By label (.loc):
print(s2.loc['b'])
```

```python
#By position (.iloc):
print(s2.iloc[1])
```

```python
#Slicing by label (inclusive of both ends!)
s4 = pd.Series([1, 2, 3, 4, 5], index=['a', 'b', 'c', 'd', 'e'])
print(s4['b':'d'])
```

```python
#Slicing by position (exclusive of end, like Python lists):
print(s4.iloc[1:4])
```

```python
#Vectorized Operations on Series
s = pd.Series([1, 2, 3, 4])
print(s * 2)

print(s + s)

print(s ** 2)

print(s > 2)
```

```python
s5 = pd.Series([1, 2, 3, 1, None, 1, 7, 8, 9])

print(s5.fillna(0))

print(s5.head(3))

print(s5.tail(5))

print(s5.describe())

print(s5.value_counts())

print(s5.sort_values(ascending=False))

print(s5.sort_index())

print(s5.isna())
```

```python
# Create a Series from this dictionary: {'Apple': 3.5, 'Banana': 2.0, 'Cherry': 5.0}
# Print the Series.
fruit_stocks = pd.Series({'Apple': 3.5, 'Banana': 2.0, 'Cherry': 5.0})

print(fruit_stocks)
```

```python
# Using the Series: s = pd.Series([100, 200, 300, 400], index=['p', 'q', 'r', 's'])
# - Access the element with label 'r' using .loc
# - Access the element at position 2 using .iloc
# - Slice from label 'q' to 's' (inclusive)
# - Slice from position 1 to 3 (exclusive end)

s = pd.Series([100, 200, 300, 400], index=['p', 'q', 'r', 's'])

print(s.loc['r'])
print(s.iloc[2])
print(s['q':'s'])
print(s.iloc[1:3])
```

```python
# Given: temps = pd.Series([32, 35, 28, 42, 38])
# Create a Boolean Series that is True where temperature > 35.
# Use that Boolean Series to filter the original Series (show only temps > 35).
temps = pd.Series([32, 35, 28, 42, 38])

print(temps > 35)

print(temps[temps > 35])
```

```python
# Create a Series: s = pd.Series([10, None, 30, None, 50])
# - Count how many missing values it has.
# - Fill missing values with the mean of the non‑missing values.
# Print the filled Series.
s = pd.Series([10, None, 30, None, 50])
print(s.isna().sum())
mean = s.mean()
print(s.fillna(mean))
```

```python
# Given: grades = pd.Series([88, 92, 79, 85, 95], index=['Ali', 'Sara', 'Omar', 'Lina', 'Zain'])
# - Sort the Series by values (descending).
# - Print summary statistics using .describe().
# - Who has the highest grade? (Print the index label of the max value.)
grades = pd.Series([88, 92, 79, 85, 95], index=['Ali', 'Sara', 'Omar', 'Lina', 'Zain'])

print(grades.sort_values(ascending=False))
print(grades.describe())
print(grades.idxmax())
```
