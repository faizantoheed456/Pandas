# 📊 Day 1, Topic 1 — What is Pandas and Why Use It?

Pandas is Python's most important **data analysis library**. Think of it as a supercharged spreadsheet that lives inside Python.

## Why Pandas?
| Problem | Pure Python/NumPy | Pandas Solution |
|---------|-------------------|-----------------|
| Tabular data with mixed types | Awkward dicts/lists | `DataFrame` — columns can have different dtypes |
| Named columns & labeled rows | Manual indexing | Built-in `Index` objects |
| Reading CSV/Excel files | Manual parsing | `pd.read_csv()`, `pd.read_excel()` |
| Grouping, aggregating | Nested loops | `.groupby()`, `.agg()` |
| Missing values | Manual `None` checks | `NaN`-aware operations everywhere |

## Two core data structures
- **`pd.Series`** — a 1-D labeled array (like a single column)
- **`pd.DataFrame`** — a 2-D labeled table (rows × columns)

Both sit on top of NumPy arrays, so they inherit all of NumPy's speed while adding label-based indexing and rich data tools.


````python
#Day 1 Topic 1: What is Pandas and Why Use It
import pandas as pd
import numpy as np
````

## `pd.Series` — The 1-D Labeled Array

A Series is like a Python list, but every element has an **index label** (not just a position number).

```python
pd.Series(data, index=labels)
```

- `data` — list, dict, scalar, or NumPy array
- `index` — optional; defaults to 0, 1, 2, …

The index lets you access elements by **name** instead of position.


````python
#Series
s = pd.Series([1, 2, 3, 4], index=['a', 'b', 'c', 'd'])
print(s)
````

## `pd.DataFrame` — The 2-D Table

A DataFrame is a collection of Series sharing the same index — essentially a table with:
- **Columns** — each is a Series (same dtype per column)
- **Index (rows)** — labels for each row (default: 0, 1, 2, …)

The most common way to create one is from a **dictionary of lists**, where keys become column names.


````python
#Dataframe
data = {
    'Name': ['Faizan', 'Ali', 'Rohan'],
    'Age': [18, 19, 22],
    'Salary': [50000.0, 25000.0, 10000.0] 
}

p = pd.DataFrame(data)
print(p)
````

## The Pandas–NumPy Bridge

Pandas is built *on top of* NumPy. Every column in a DataFrame is backed by a NumPy array. You can access it with `.values`:

```python
series.values        # → numpy.ndarray
dataframe['col'].values  # → numpy.ndarray
```

This means you can pass Pandas columns directly to any NumPy function, SciPy routine, or scikit-learn model.


````python
print(p['Age'].values)
print(type(p['Age'].values))
````

## Vectorized Operations — No Loops Needed

Just like NumPy, Pandas operations work element-wise on entire columns without Python loops:

```python
df['Salary'] *= 1.1   # gives every salary a 10% raise — all at once
df['Age'] > 20        # returns a boolean Series — True/False per row
```

These operations run in compiled C code and are much faster than equivalent Python loops.


````python
p['Salary'] *= 1.1
````

````python
p['Age'] > 20
````

### Task 1 — Series from a NumPy Array
Create 5 random integers with `np.random.randint`, wrap in a DataFrame with custom index labels `['a','b','c','d','e']`.


````python
#Task 1: Create a Series from a NumPy Array
import pandas as pd
import numpy as np

# Create a NumPy array of 5 random integers between 1 and 100
arr = np.random.randint(1, 101, size=5)

df = pd.DataFrame(arr, index=['a', 'b', 'c', 'd', 'e'])
print(df)
````

### Task 2 — DataFrame from a Dictionary
Build a product inventory table from a dict. Notice that each dict key becomes a column name automatically.


````python
# Create a DataFrame from this dictionary:
data = {
    'Product': ['Laptop', 'Mouse', 'Keyboard', 'Monitor'],
    'Price': [1200, 25, 75, 300],
    'Quantity': [5, 50, 30, 10]
}

# Print the DataFrame
df = pd.DataFrame(data)
print(df)
````

### Task 3 — Extract a Column as NumPy Array
`df['Price'].values` strips away the Pandas wrapper and gives you a raw `numpy.ndarray`.  
Use this when passing data to libraries that don't understand Pandas (e.g., raw scikit-learn, OpenCV).


````python
# Using the DataFrame from Task 2, extract the 'Price' column as a NumPy array.
# Print the array and verify its type is numpy.ndarray.
print(df['Price'].values)
print(type(df['Price'].values))
````

### Task 4 — Series with Float Index Labels
Demonstrates that index labels can be any hashable type — strings like `'A'`, `'B'` are very common.  
Note: here a DataFrame is created instead of a Series, which still works but wraps the data in a single-column table.


````python
import pandas as pd
import numpy as np

# Create a NumPy array of 4 float values: 10.5, 20.3, 30.8, 40.1
arr = np.array([10.5, 20.3, 30.8, 40.1])

# Create a Pandas Series from this array with index labels: 'A', 'B', 'C', 'D'
# Assign it to a variable 's' and print it.
s = pd.DataFrame(arr, index=['A', 'B', 'C', 'D'])
print(s)
````

### Task 5 — City Population DataFrame
Straightforward dict → DataFrame construction. Real-world datasets almost always start this way.


````python
# Given this dictionary:
data = {
    'City': ['Lahore', 'Karachi', 'Islamabad'],
    'Population': [11126285, 14910352, 1014825]
}

# Create a DataFrame from this dictionary. Print it.
df = pd.DataFrame(data)
print(df)
````

### Task 6 — Vectorized Division on a Column
Extracting a column gives you a `pd.Series`. Dividing by a scalar is vectorized — it applies to every element and returns a new Series. The original DataFrame is unchanged.


````python
# Using the DataFrame you just created, extract the 'Population' column.
# Then use a NumPy ufunc on it—for example, divide every value by 1,000,000 to get population in millions.
# Print the result. It should still be a Pandas Series.
pop_series = df['Population']           # This is a Series
pop_millions = pop_series / 1_000_000   # Vectorized division, returns a Series
print(pop_millions)
print(type(pop_millions))               # <class 'pandas.core.series.Series'>
````

### Task 7 — NumPy Array → DataFrame
`np.zeros((2,3))` creates a 2-D NumPy array. Wrapping it in `pd.DataFrame` with explicit `columns` and `index` gives it human-readable labels — useful for initializing result tables.


````python
# Create a 2x3 NumPy array of zeros.
# Convert it to a DataFrame with:
#   - columns named: 'X', 'Y', 'Z'
#   - index named: 'first', 'second'
# Print the DataFrame.
arr = np.zeros((2, 3))

df = pd.DataFrame(arr, columns=['X', 'Y', 'Z'], index=['first', 'second'])
print(df)
````
