```python
#Day 1, Topic 3: The DataFrame Object – Deep Dive
import pandas as pd
import numpy as np
```

```python
#From a dictionary of lists (most common):

data = {
    'Name': ['Faizan', 'Ali', 'Rohan'],
    'Age': [22, 18, 20],
    'Salary': [50000, 67000, 32000]
}

df = pd.DataFrame(data)
print(df)
```

```python
#From a list of dictionaries (each dict = one row):

data = [
    {'Name': 'Faizan', 'Age': 18, 'Salary': 40000},
    {'Name': 'Rohan', 'Age': 19, 'Salary': 60000},
    {'Name': 'Ali', 'Age': 17, 'Salary': 50000},
]

df = pd.DataFrame(data)

print(df)
```

```python
#From a list of lists with explicit column names:

data = [
    ['Faizan', 18, 40000],
    ['Usman', 19, 50000],
    ['Rohan', 17, 30000],
]

df = pd.DataFrame(data, columns=['Name', 'Age', 'Salary'])


print(df)
```

```python
#From a dictionary of Series (useful when you already have Series):
s1 = pd.Series(['Faizan', 'Arslan', 'Rohan'], name='Name')
s2 = pd.Series([18, 17, 19], name='Age')
s3 = pd.Series([40000, 50000, 43000], name='Salary')

df = pd.DataFrame({'Name': s1, 'Age': s2, 'Salary': s3})
print(df)
```

```python
#Practical Activity: Topic 3
import pandas as pd

data = {
    'Product': ['Laptop', 'Mouse', 'Keyboard', 'Monitor', 'Headphones'],
    'Category': ['Electronics', 'Accessories', 'Accessories', 'Electronics', 'Accessories'],
    'Price': [1200, 25, 75, 300, 80],
    'Quantity': [5, 50, 30, 10, 40],
    'Rating': [4.5, 4.2, 4.7, 4.3, 4.8]
}
df = pd.DataFrame(data)
```

```python
df
```

```python
"""Task 1: Basic Inspection
Print the first 3 rows.
Print the shape (rows, columns).
Print the column names.
Print the data types of all columns.
Print summary statistics for numeric columns only."""
print("\nThe first 3 rows from data are given below:\n", df.head(3))
print("\nThe shape of dataset is:", df.shape)
print("\nDisplaying the column names from dataset: ", df.columns)
print("\nStats for numeric columns:\n", df.describe())
```

```python
df
```

```python
"""Task 2: Column Operations
Create a new column 'Revenue' which is Price * Quantity.
Increase all prices by 5% (update the 'Price' column in‑place).
Drop the 'Rating' column and assign the result back to df."""
df['Revenue'] = df['Price'] * df['Quantity']
df['Price'] *= 1.05
df = df.drop('Rating', axis=1)
print("\nDataset after adding a new column, increase an existing column and removing the rating column:\n", df)
```

```python
df
```

```python
"""Task 3: Boolean Filtering
Create a Boolean mask for products where Category is 'Accessories'.
Use that mask to display only accessory products.
Display products where Price is greater than 100 and Quantity is less than 20."""
mask = (df['Category'] == 'Accessories')
accessory_products = df[mask]
print("\nAccessory product data set:\n", accessory_products)
price_mask = (df['Price'] > 100) & (df['Quantity'] < 20)
print("\nDisplaing products where Price is greater than 100 and Quantity is less than 20:\n", df[price_mask])
```

```python
df
```

```python
"""Task 4: Access with .loc and .iloc
Using .loc, select the row with index label 2 and the column 'Product'.
Using .iloc, select the first row and the last column (position‑based).
Using .loc, select rows 0 through 2 (inclusive) and columns 'Product' and 'Price'.
"""
print("Selecting the row with index label 2 and the column 'Product': ", df.loc[2, 'Product'])
print("Selecting the first row and the last column: ", df.iloc[0, 4])
print("\nSelecting rows 0 through 2 (inclusive) and columns 'Product' and 'Price':\n",df.loc[0:2, ['Product']])
```
