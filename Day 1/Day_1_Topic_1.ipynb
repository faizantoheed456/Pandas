```python
import pandas as pd
import numpy as np

# Create a NumPy array of 4 float values: 10.5, 20.3, 30.8, 40.1
arr = np.array([10.5, 20.3, 30.8, 40.1])

# Create a Pandas Series from this array with index labels: 'A', 'B', 'C', 'D'
# Assign it to a variable 's' and print it.
s = pd.DataFrame(arr, index=['A', 'B', 'C', 'D'])
print(s)
```

```python
# Given this dictionary:
data = {
    'City': ['Lahore', 'Karachi', 'Islamabad'],
    'Population': [11126285, 14910352, 1014825]
}

# Create a DataFrame from this dictionary. Print it.
df = pd.DataFrame(data)
print(df)
```

```python
# Using the DataFrame you just created, extract the 'Population' column.
# Then use a NumPy ufunc on it—for example, divide every value by 1,000,000 to get population in millions.
# Print the result. It should still be a Pandas Series.
pop_series = df['Population']           # This is a Series
pop_millions = pop_series / 1_000_000   # Vectorized division, returns a Series
print(pop_millions)
print(type(pop_millions))               # <class 'pandas.core.series.Series'>
```

```python
# Create a 2x3 NumPy array of zeros.
# Convert it to a DataFrame with:
#   - columns named: 'X', 'Y', 'Z'
#   - index named: 'first', 'second'
# Print the DataFrame.
arr = np.zeros((2, 3))

df = pd.DataFrame(arr, columns=['X', 'Y', 'Z'], index=['first', 'second'])
print(df)
```
