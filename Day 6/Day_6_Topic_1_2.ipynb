```python
#Day 6, Topic 1: Converting to Datetime

import pandas as pd
```

```python
#Basic pd.to_datetime()
dates = pd.Series(['2024-01-01', '2024-02-01', '2024-03-01'])
print(dates.dtype)

datetime_dates = pd.to_datetime(dates)
print(datetime_dates.dtype)
datetime_dates
```

```python
#Handling Different Date Formats
dates_uk = pd.Series(['01/01/2024', '02/03/2025']) #DD/MM/YYYY
datetime_dates = pd.to_datetime(dates_uk, format='%d/%m/%Y')
datetime_dates
```

```python
#The errors Parameter
messy_date = pd.Series(['2024/02/01', 'Not a date', '2024/03/03'])
clean_date = pd.to_datetime(messy_date, errors='coerce')
clean_date
```

```python
#Converting DataFrame Columns
df = pd.DataFrame({
    'Name': ['Faizan', 'Ali'],
    'Join_Date': ['2023-01-23', '2024-11-9']
})

df['Join_Date'] = pd.to_datetime(df['Join_Date'])
df
```

```python
#Working with Timestamps
single_date = pd.to_datetime('2026-04-22')

print(single_date.year)
print(single_date.month)
print(single_date.day)
print(single_date.dayofweek + 1)
```

```python
#Practice Activity: Datetime Conversion
import pandas as pd

data = {
    'Event': ['Conference', 'Workshop', 'Seminar', 'Meetup'],
    'Date_Str': ['2024-03-15', '2024/04/20', 'Invalid', '15-05-2024']
}
df = pd.DataFrame(data)
df
```

```python
"""Task 1: Convert the Date_Str column to datetime using pd.to_datetime() with errors='coerce'. 
Display the DataFrame and note which row becomes NaT."""

result = pd.to_datetime(df['Date_Str'], errors='coerce')

result
```

```python

"""Task 2: The valid dates are in mixed formats. Use the format parameter to correctly 
parse the third row ('15-05-2024' as '%d-%m-%Y') while still using errors='coerce'. """

df['Date_Str'] = pd.to_datetime(df['Date_Str'], format='mixed', errors='coerce')
df
```

```python
#Task 3: Create a new column Year that extracts the year from the converted datetime column.

df['Year'] = df['Date_Str'].apply(lambda x: x.year if pd.notnull(x) else None)

df
```

```python
#Task 4 (Bonus): Convert a single string '2024-12-25 14:30:00' to a Timestamp. Then extract the hour and minute from it.

single_datetime = pd.to_datetime('2024-12-25 14:30:00')
print(single_datetime.hour)
print(single_datetime.minute)
```

```python
dates = pd.Series(pd.to_datetime(['2026-01-05', '2026-01-06', '2026-01-07']))

print(dates.dt.year)
print(dates.dt.month)
print(dates.dt.day)
print(dates.dt.hour)
print(dates.dt.minute)
print(dates.dt.second)
```

```python
#Day of Week, Quarter, and More
print(dates.dt.dayofweek)
print(dates.dt.day_name())
print(dates.dt.quarter)
print(dates.dt.is_month_start)
print(dates.dt.is_month_end)
print(dates.dt.is_leap_year)
print(dates.dt.isocalendar())
```

```python
#Practice Activity: .dt Accessor
import pandas as pd
import numpy as np

df = pd.read_csv('titanic.csv')

departure = pd.to_datetime('1912-04-15')
df['Booking_Date'] = departure - pd.Timedelta(days=30)
np.random.seed(42)
df['Booking_Date'] = df['Booking_Date'] + pd.to_timedelta(np.random.randint(-5, 5, size=len(df)), unit='D')
```

```python
#Task 1: Create a new column Booking_Quarter containing the quarter (1–4) of the Booking_Date. Use .dt.quarter.

df['Booking_Quarter'] = df['Booking_Date'].dt.quarter
df[['Booking_Date', 'Booking_Quarter']].head(10)
```

```python
"""Task 2: Create a new column Booking_Weekend that is True if the booking day is Saturday or Sunday, 
False otherwise. Use .dt.dayofweek (Monday=0, Sunday=6)."""

df['Booking_Weekend'] = df['Booking_Date'].dt.dayofweek >= 5
df[['Booking_Date', 'Booking_Weekend']].head()
```

```python
#Task 3: Count how many bookings were made in each month. Use .dt.month and .value_counts().

df['Booking_Date'].dt.month.value_counts()
```

```python
#Task 4: Filter the DataFrame to show only bookings made on the first day of the month. Use .dt.is_month_start.

month_start = df[df['Booking_Date'].dt.is_month_start]
len(month_start)
```

```python
"""Task 5 (Bonus): Format the Booking_Date as a string like "Monday, 16 March 1912" using 
.dt.strftime() with format '%A, %d %B %Y'. Display the first 5 rows."""

df['format_date'] = df['Booking_Date'].dt.strftime('%A %d %B %Y')

df[['Booking_Date', 'format_date']].head()
```
