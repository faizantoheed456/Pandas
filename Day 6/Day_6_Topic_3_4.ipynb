```python
import pandas as pd
import numpy as np

df = pd.read_csv('titanic.csv')
# Simulate booking date and departure date
departure = pd.to_datetime('1912-04-15')
np.random.seed(42)
df['Booking_Date'] = departure - pd.Timedelta(days=30) + pd.to_timedelta(np.random.randint(-5, 5, size=len(df)), unit='D')
df['Departure_Date'] = departure
```

```python
#Subtracting Dates – Time Differences
df['Days_Ahead'] = df['Departure_Date'] - df['Booking_Date']
df[['Booking_Date', 'Departure_Date', 'Days_Ahead']].head()
```

```python
#Creating a Timedelta from a Column
df['Offset_Days'] = np.random.randint(1, 10, size=len(df))
df['New_Date'] = df['Booking_Date'] + pd.to_timedelta(df['Offset_Days'], unit='D')
df[['Booking_Date', 'New_Date']].head()
```

```python
#Practice Activity: Date Arithmetic
import pandas as pd
import numpy as np


df = pd.read_csv('titanic.csv')

departure = pd.to_datetime('1912-04-15')
np.random.seed(42)
df['Booking_Date'] = departure - pd.Timedelta(days=30) + pd.to_timedelta(np.random.randint(-10, 10, size=len(df)), unit='D')
df['Departure_Date'] = departure
```

```python
"""Task 1: Create a new column Days_Before that is the number of days between Booking_Date 
and Departure_Date as an integer (use .dt.days)."""

df['Days_Before'] = df['Departure_Date'] - df['Booking_Date']
df['Days_Before'] = df['Days_Before'].dt.days
df[['Booking_Date', 'Departure_Date', 'Days_Before']].head()
```

```python
#Task 2: Create a column Booking_2Weeks_Later that is Booking_Date plus 14 days.

df['Booking_2Weeks_Later'] = df['Booking_Date'] + pd.Timedelta(days=14)
df[['Booking_Date', 'Booking_2Weeks_Later']].head()
```

```python
"""Task 3: Create a Boolean column Is_Early_Bird that is True if the booking was made more than 35 
days before departure, False otherwise."""
# 1. Calculate the difference (this creates a Timedelta Series)
# 2. Compare it to a Timedelta of 35 days
df['Is_Early_Bird'] = (df['Departure_Date'] - df['Booking_Date']) > pd.Timedelta(days=35)

# Display a sample of the results
print(df[['Booking_Date', 'Departure_Date', 'Is_Early_Bird']].head())

# Check how many passengers qualified
print(f"\nNumber of Early Birds: {df['Is_Early_Bird'].sum()}")
```

```python
"""Task 4: Filter the DataFrame to show only passengers who booked exactly 30 days before departure. 
How many are there? (Hint: compare Days_Before == 30.)"""

exact_pass = df[df['Days_Before'] == 30]

len(exact_pass)
```

```python
"""Task 5 (Bonus): Calculate the average number of days before departure that passengers booked, 
grouped by Pclass. Which class booked the earliest on average?"""

avg_days = df.groupby('Pclass')['Days_Before'].mean()
avg_days
```

```python
#Day 6, Topic 4: Resampling Time Series
import pandas as pd
import numpy as np

df = pd.read_csv('titanic.csv')

# Simulate a 'Booking_Date' column for each passenger (already done in earlier topics)
departure = pd.to_datetime('1912-04-15')
np.random.seed(42)
df['Booking_Date'] = departure - pd.Timedelta(days=30) + pd.to_timedelta(np.random.randint(-10, 10, size=len(df)), unit='D')
```

```python
daily_sampling = df.resample('D', on='Booking_Date')['PassengerId'].count()

daily_sampling.head(10)
```

```python
#Example: Weekly bookings
weekly_booking = df.resample('W-MON', on='Booking_Date')['PassengerId'].count()
weekly_booking
```

```python
#Multiple Aggregations
fare_stats = df.resample('W', on='Booking_Date')['Fare'].agg(['sum', 'mean', 'count'])
fare_stats
```

```python
#Resampling with a Datetime Index
df_indexed = df.set_index('Booking_Date')
monthly_avg_fare = df_indexed['Fare'].resample('M').mean()
monthly_avg_fare
```

```python
# Create a small daily series
daily = df.resample('D', on='Booking_Date')['Fare'].mean()
# Upsample to 12-hour intervals
upsampled = daily.resample('12H').ffill()
print(upsampled.head(10))
```

```python
#Practice Activity: Resampling
import pandas as pd
import numpy as np

df = pd.read_csv('titanic.csv')

departure = pd.to_datetime('1912-04-15')
np.random.seed(42)
df['Booking_Date'] = departure - pd.Timedelta(days=30) + pd.to_timedelta(np.random.randint(-10, 10, size=len(df)), unit='D')
```

```python
"""Task 1: Resample the data to weekly frequency (ending on Sunday, 'W') 
and compute the total Fare collected each week. Display the result."""

weekly = df.resample('W', on='Booking_Date')['Fare'].sum()
weekly
```

```python
"""Task 2: Resample to monthly frequency ('M') and compute both the average
 Fare and number of bookings in a single .agg() call."""

montly = df.resample('ME', on='Booking_Date')['Fare', 'PassengerId'].agg({'Fare': 'mean', 'PassengerId': 'count'})
montly
```

```python
"""Task 3: Set Booking_Date as the index, then resample to daily frequency and 
count how many passengers booked each day. Fill any missing days with 0 
(hint: use .resample('D').count() and then .fillna(0))."""

df_indexed = df.set_index('Booking_Date')
daily = df_indexed.resample('D')['PassengerId'].count().fillna(0)

daily.head(10)
```

```python
"""Task 4 (Bonus): Compute a 7‑day rolling average of daily booking counts. 
(We'll preview rolling windows next, but try: daily_bookings.rolling(7).mean())."""

rolling_7d = daily.rolling(window=7).mean()
rolling_7d.head(10)
```
