```python
#Day 5, Topic 1: Concatenating with pd.concat()
import pandas as pd

df = pd.read_csv('titanic.csv')
```

```python
#Vertical Concatenation (axis=0 – The Default)
df1 = df.iloc[:5]
df2 = df.iloc[5:10]

result = pd.concat([df1, df2], axis=0)
print(result.shape)
result
```

```python
result_clean = pd.concat([df1, df2], ignore_index=True)

result_clean.index
```

```python
#Horizontal Concatenation (axis=1)
df_left = df[['Name', 'Age', 'Sex']].iloc[:5]
df_right = df[['Fare', 'Embarked']].iloc[:5]

result = pd.concat([df_left, df_right], axis=1)
result
```

```python
#Adding Keys for MultiIndex (keys Parameter)
southampton = df[df['Embarked'] == 'S'].head(3)
cherbourg = df[df['Embarked'] == 'C'].head(3)

result = pd.concat([southampton, cherbourg], keys=['southampton', 'cherbourg'])

result
```

```python
#Practice Activity: pd.concat()
import pandas as pd
df = pd.read_csv('titanic.csv')
# Create three separate DataFrames
df_first = df[df['Pclass'] == 1].iloc[:3]   # First 3 rows of 1st class
df_second = df[df['Pclass'] == 2].iloc[:3]  # First 3 rows of 2nd class
df_third = df[df['Pclass'] == 3].iloc[:3]   # First 3 rows of 3rd class
```

```python
# Task 1: Vertically concatenate df_first, df_second, df_third in that order.
# Use ignore_index=True for a clean index.
# What is the shape of the result?
result = pd.concat([df_first, df_second, df_third], axis=0, ignore_index=True)
result.shape
```

```python
# Task 2: Concatenate the same three DataFrames vertically, but this time use 
# the keys parameter with labels ['First', 'Second', 'Third'].
# Use .loc to select only the rows from the 'Second' group.
concat = pd.concat([df_first, df_second, df_third], axis=0, keys=['First', 'Second', 'Third'])
concat.loc['Second']
```

```python
# Task 3: Split df into two horizontal halves:
# left_half: columns 'Name', 'Sex', 'Age', 'SibSp', 'Parch', 'Ticket'
# right_half: columns 'Fare', 'Cabin', 'Embarked'
# Take the first 5 rows of each and concatenate them horizontally (axis=1).
# Verify that the rows align correctly.
left_cols = ['Name', 'Sex', 'Age', 'SibSp', 'Parch', 'Ticket']
right_cols = ['Fare', 'Cabin', 'Embarked']

df_1 = df.loc[:4, left_cols]
df_2 = df.loc[:4, right_cols]

result = pd.concat([df_1, df_2], axis=1)
result
```

```python
# Task 4 (Bonus): Create a function that takes a list of DataFrames and returns 
# a single DataFrame concatenated vertically, but only keeps columns that are 
# present in ALL of the input DataFrames

def concat_dataframes(df_1, df_2):
    return pd.concat([df_1, df_2], axis=0, join='inner')

df_1 = df.iloc[:5]
df_2 = df.iloc[5: 10]

result = concat_dataframes(df_1=df_1, df_2=df_2)
result
```

```python
#Day 5, Topic 2: Merging with pd.merge()

import pandas as pd

df = pd.read_csv('titanic.csv')
```

```python
#The Two DataFrames We'll Join
passengers = df[['PassengerId', 'Name', 'Sex', 'Age']].copy()

tickets = df[['PassengerId', 'Pclass', 'Fare', 'Embarked']].copy()

tickets = tickets.drop([2, 5, 10])

print(passengers.shape)
print(tickets.shape)
```

```python
#Inner Join (how='inner')
inner_joined = pd.merge(passengers, tickets, on='PassengerId', how='inner')
print(inner_joined.shape)
inner_joined.head()
```

```python
#Left Join (how='left')
left_joined = pd.merge(passengers, tickets, on='PassengerId', how='left')
print(left_joined.shape)
left_joined.head()
```

```python
#Right Join (how='right')
right_joined = pd.merge(passengers, tickets, on='PassengerId', how='right')
print(right_joined.shape)
right_joined.head()
```

```python
#Outer Join (how='outer')
outer_joined = pd.merge(passengers, tickets, on='PassengerId', how='outer')
print(outer_joined.shape)
```

```python
#Merging on Different Column Names
tickets_renamed = tickets.rename(columns={'PassengerId': 'PID'})

merged = pd.merge(passengers, tickets_renamed, left_on='PassengerId', right_on='PID', how='inner')
merged.head()
```

```python
#Merging on Multiple Columns
# Example: Create a surrogate key
passengers['Deck'] = passengers['Name'].str[0]
tickets['Deck'] = tickets['Embarked']  # Not logical, just for demo

merged_multi = pd.merge(passengers, tickets, on=['PassengerId', 'Deck'], how='inner')
merged_multi.head()
```

```python
#Handling Duplicate Column Names
merged = pd.merge(passengers, tickets, on='PassengerId', suffixes=('_Passenger', '_Ticket'))

merged.columns
```

```python
#Practice Activity: pd.merge()
import pandas as pd

df = pd.read_csv('titanic.csv')

#The Two DataFrames We'll Join
passengers = df[['PassengerId', 'Name', 'Sex', 'Age']].copy()

tickets = df[['PassengerId', 'Pclass', 'Fare', 'Embarked']].copy()

tickets = tickets.drop([2, 5, 10])

print(passengers.shape)
print(tickets.shape)
```

```python
#Task 1: Perform an inner join on PassengerId. How many rows are in the result? Why is it less than 891?

inner_joined = pd.merge(passengers, tickets, on='PassengerId', how='inner')

len(inner_joined)
```

```python
"""Task 2: Perform a left join on PassengerId. Find a passenger who has missing ticket information. 
Display their PassengerId, Name, and Fare."""

left_joined = pd.merge(passengers, tickets, on='PassengerId', how='left')

missing_tickets = left_joined[left_joined['Fare'].isna()]

if not missing_tickets.empty:
    result = missing_tickets[['PassengerId', 'Name', 'Fare']].head(1)
    print(result)
else:
    print("No passengers found with missing ticket information.")
```

```python
"""Task 3: Rename the PassengerId column in tickets to 'PID'. Then perform an inner join using left_on='PassengerId' 
and right_on='PID'. Drop the duplicate 'PID' column from the result.
"""

tickets_renamed = tickets.rename(columns={'PassengerId': 'PID'})

inner_merged = pd.merge(passengers, tickets_renamed, left_on='PassengerId', right_on='PID', how='inner')
inner_merged.drop('PID', axis=1)
```

```python
"""Task 4: Suppose you want to keep only passengers who DO NOT have a matching ticket record. 
How can you achieve this using a merge? (Hint: how='left' and then filter for NaN, 
or use the indicator=True parameter.)"""

left_joined = pd.merge(passengers, tickets, on='PassengerId', how='left')
no_ticket = left_joined[left_joined['Fare'].isna()]
no_ticket[['PassengerId', 'Name']].head()
```

```python
"""Task 5 (Bonus): Use pd.merge() with indicator=True to add a '_merge' column showing the 
source of each row ('both', 'left_only', 'right_only'). Then count how many rows came from each source."""

merged_indicator = pd.merge(passengers, tickets, on='PassengerId', how='outer', indicator=True)
merged_indicator['_merge'].value_counts()
```
