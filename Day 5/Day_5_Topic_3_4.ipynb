```python
#Day 5, Topic 3: Joining with .join()

import pandas as pd

df = pd.read_csv('titanic.csv')
```

```python
#Basic Index‑Based Join
passengers = df[['Name', 'Sex', 'Age']].set_index(df['PassengerId'])
tickets = df[['Pclass', 'Fare', 'Embarked']].set_index(df['PassengerId'])

result = passengers.join(tickets)

result.head()
```

```python
#Specifying Join Type with how
inner_joined = passengers.join(tickets, how='inner')
inner_joined.shape
```

```python
#Joining on a Column Instead of Index
passenger_reset = passengers.reset_index()
ticket_reset = tickets.reset_index()

joined = passenger_reset.join(ticket_reset.set_index('PassengerId'), on='PassengerId')

joined.head()
```

```python
#Handling Overlapping Column Names
#passengers['Score'] = 100
#tickets['Score'] = 200

#joined = passengers.join(tickets, lsuffix='_passenger', rsuffix='_ticket')
#joined.columns
```

```python
#Joining Multiple DataFrames at Once
survival = df[['Survived']].set_index(df['PassengerId'])

result = passengers.join([tickets, survival])
result.head()
```

```python
#Practice Activity: .join()
import pandas as pd
df = pd.read_csv('titanic.csv')
# Create three DataFrames, each indexed by 'PassengerId'
df1 = df[['Name', 'Sex']].set_index(df['PassengerId'])
df2 = df[['Age', 'Fare']].set_index(df['PassengerId'])
df3 = df[['Survived', 'Pclass']].set_index(df['PassengerId'])
```

```python
#Task 1: Use .join() to combine df1 and df2 (left join). Display the first 3 rows.

result = df1.join(df2, how='left')
result.head(3)
```

```python
#Task 2: Join all three DataFrames (df1, df2, df3) in a single .join() call. Use an inner join.

result = df1.join([df2, df3], how='inner')
result.head()
```

```python
"""Task 3: Create a new DataFrame df4 that has the same index as df1 but contains a column 
'Deck' (first letter of Cabin). Then join df4 to df1 and specify rsuffix='_Deck' 
to handle any duplicate columns (though there shouldn't be any, it's good practice). Display the result."""

df4 = pd.DataFrame({'Deck': df['Cabin'].str[0]}, index=df1.index)

result = df1.join(df4, rsuffix='_Deck')

result.head()
```

```python
"""Task 4 (Bonus): Reset the index of df1 and df2 so that PassengerId becomes a column. 
Now use .join() with the on parameter to join df2 
(after setting its index back to PassengerId) to df1. This demonstrates the column‑on‑index join."""

df1_reset = df1.reset_index()
df2_reset = df2.reset_index()

result = df1_reset.join(df2_reset.set_index('PassengerId'), on='PassengerId')
result.head()
```

```python
#Day 5, Topic 4: Handling Duplicate Column Names

import pandas as pd
```

```python
#Default Suffixes (_x and _y)
left = pd.DataFrame({
    'ID': [1, 2, 3],
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Score': [85, 90, 78]
})

right = pd.DataFrame({
    'ID': [1, 2, 3],
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Score': [88, 92, 80]
})

merged = pd.merge(left, right, on='ID')

merged
```

```python
#Customizing Suffixes with suffixes
merged = pd.merge(left, right, on='ID', suffixes=('_Left', '_Right'))
merged.columns.tolist()
```

```python
#With .join():

left_joined = left.set_index('ID').join(right.set_index('ID'), lsuffix='_L', rsuffix='_R')
print(left_joined.columns.tolist())
```

```python
#Real Titanic Example
df = pd.read_csv('titanic.csv')
# Two DataFrames with overlapping 'Name' and 'Age' columns
passengers = df[['PassengerId', 'Name', 'Age', 'Sex']].copy()
age_backup = df[['PassengerId', 'Name', 'Age']].copy()
age_backup['Age'] = age_backup['Age'].fillna(30)  # Filled ages

merged = pd.merge(passengers, age_backup, on='PassengerId', suffixes=('_Original', '_Filled'))
merged.columns.tolist()
```

```python
#Preventing Duplicate Columns Before Merge
age_backup_renamed = age_backup.rename(columns={'Name': 'Name_backup', 'Age': 'Age_filled'})
merged = pd.merge(passengers, age_backup_renamed, on='PassengerId')

merged.columns.tolist()
```

```python
#Practice Activity: Handling Duplicate Columns
import pandas as pd
df = pd.read_csv('titanic.csv')
# Create two overlapping DataFrames
df_a = df[['PassengerId', 'Name', 'Age', 'Fare']].iloc[:10].copy()
df_b = df[['PassengerId', 'Name', 'Pclass', 'Survived']].iloc[:10].copy()
```

```python
#Task 1: Merge df_a and df_b on 'PassengerId'. Observe the default column names. Print the column list.

merged = pd.merge(df_a, df_b, on='PassengerId')
merged.columns.tolist()
```

```python
#Task 2: Repeat the merge, but use suffixes=('_A', '_B') to clearly label the source DataFrame. Print the column list.

merged = pd.merge(df_a, df_b, on='PassengerId', suffixes=('_A', '_B'))
merged.columns.tolist()
```

```python
#Task 3: Drop the duplicate Name column that came from df_b (i.e., 'Name_B'). Verify the remaining columns.
merged = merged.drop(columns=['Name_B'])
merged.columns.tolist()
```

```python
"""Task 4 (Bonus): Rename the columns in df_b before merging so that no duplicate column names exist. 
Specifically, rename 'Name' to 'Name_FromB' and 'Pclass' to 'Class'. Then perform the merge. 
This is the cleanest production approach."""

df_B_renamed = df_b.rename(columns={'Name': 'Name_FromB', 'Pclass': 'Class'})
merged = pd.merge(df_a, df_B_renamed, on='PassengerId')
merged.columns.tolist()
```
