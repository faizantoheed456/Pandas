````python
#Day 3, Topic 7: Renaming Columns
import pandas as pd
df = pd.read_csv('titanic.csv')
````

## Topic 7 — Renaming Columns

Good column names are lowercase, use underscores (not spaces), and are self-explanatory.  
Renaming is always one of the first steps in a data cleaning pipeline.

### The renaming toolkit

| Method | Use |
|--------|-----|
| `df.rename(columns={'old': 'new'})` | Rename specific columns |
| `df.columns = [list of names]` | Replace all column names at once |
| `df.columns.str.lower()` | Vectorized: lowercase all names |
| `df.columns.str.replace(' ', '_')` | Replace spaces with underscores |
| `df.rename(columns=str.upper)` | Apply a function to all column names |
| `df.rename(columns=lambda x: x+'_col')` | Lambda for custom transformation |
| `df.add_prefix('prefix_')` | Add prefix to all column names |
| `df.add_suffix('_suffix')` | Add suffix to all column names |


### `.rename()` — Rename Specific Columns
Pass a dictionary `{old_name: new_name}`. Only the listed columns are affected; others are unchanged.  
Always assign back: `df = df.rename(...)` (or use `inplace=True`).


````python
#.rename() – The Primary Tool
df = df.rename(columns={
    'Pclass': 'Passenger_Class',
    'SibSp': 'Siblings_Spouses'
})

df.columns.tolist()
````

### Rename All Columns at Once — Vectorized
`df.columns = df.columns.str.lower()` applies `.lower()` to every column name simultaneously.  
This is the standard first step in column standardization.


````python
#Renaming All Columns at Once

df.columns = df.columns.str.lower()
df.columns.tolist()
````

### Replace Spaces with Underscores
`df.columns.str.replace(' ', '_')` — columns with spaces can't use dot notation and are harder to reference.  
Always clean spaces before doing string-based column selection.


````python
#Replace spaces with underscores:
df.columns.str.replace(' ', '_')
````

### Apply a Function to All Column Names
`df.rename(columns=str.upper)` or `df.rename(columns=lambda x: x + '_col')` applies a function to every name.  
More flexible than hardcoding a dictionary when you want a systematic transformation.


````python
#Using a Function to Transform Names
# Convert all columns to uppercase
df = df.rename(columns=str.upper)

# Or use a lambda
df = df.rename(columns=lambda x: x.replace('_', ''))
````

### `.add_prefix()` and `.add_suffix()`
Quick way to namespace columns — useful when merging DataFrames to avoid name collisions:
```python
df_left.add_suffix('_left')
df_right.add_suffix('_right')
```


````python
#Adding Prefixes or Suffixes
df_prefixed = df.add_prefix('col_')

df_suffixed = df.add_suffix('_titanic')

````

### Full Standardization Pipeline
```python
df.columns = (df.columns
    .str.lower()                          # lowercase
    .str.replace(' ', '_')               # spaces → underscores
    .str.replace('[^a-z0-9_]', '', regex=True)  # remove special chars
)
```
Run this once on any new dataset and you'll have clean, Python-friendly column names.


````python
#Titanic Column Standardization
df.columns = (df.columns.str.lower().str.replace(' ', '_').str.replace('[^a-z0-9_]', '', regex=True))
````

## Topic 8 — Replacing Values

`.replace()` is the primary tool for correcting known bad values, encoding categories, and standardizing representations.

### `.replace()` forms

| Form | Example |
|------|---------|
| Single value | `df['Sex'].replace('male', 'M')` |
| Multiple values → same | `df['Emb'].replace(['C','Q'], 'other')` |
| Dictionary mapping | `df['Emb'].replace({'S':'Southampton', 'C':'Cherbourg'})` |
| Regex | `df['Name'].replace(r'Mr\..*', 'Adult', regex=True)` |
| Replace NaN | `df['Cabin'].replace(np.nan, 'Unknown')` |
| Whole DataFrame | `df.replace({'male': 0, 'female': 1})` |

> **`.replace()` vs `.map()`:**  
> `.replace()` — leaves unmatched values unchanged  
> `.map()` — sets unmatched values to `NaN`  
> Use `.replace()` when only changing some values; `.map()` when remapping all values.


### Replace a Single Value
`df['Sex'].replace('male', 'M')` — only 'male' is changed; 'female' remains.  
Returns a new Series — `df` is not modified unless you reassign.


````python
#Replace a Single Value
df['Sex'] = df['Sex'].replace('male', 'M')
df['Sex'].value_counts()
````

### Replace Multiple Values → Same Replacement
`df['Embarked'].replace(['C', 'Q'], 'other')` — both 'C' and 'Q' become 'other' in one call.


````python
#Replace Multiple Values
df['Embarked'] = df['Embarked'].replace(['C', 'Q'], 'other')
df['Embarked'].value_counts()
````

### Dictionary Mapping — The Most Readable Form
```python
{'S': 'Southampton', 'C': 'Cherbourg', 'Q': 'Queenstown'}
```
Each key is replaced by its value. Unmatched values stay unchanged.  
This is the preferred form for categorical encoding/decoding.


````python
#Replace Using a Dictionary

embark_mapping = {
    'S': 'Southampton', 
    'C': 'Cherbourg',
    'Q': 'Queenstown'
}

df['Embarked'] = df['Embarked'].replace(embark_mapping)
df['Embarked'].value_counts()
````

### Replace with Regex
`regex=True` enables Python regex patterns in the search string.  
`r'.*Mr\.|.*Mrs.*'` — any string containing 'Mr.' or 'Mrs' → 'Adult'.  
⚠️ Regex replace can be powerful but unpredictable — always verify on a small sample first.


````python
#Replace Using Regular Expressions
df['Name'] = df['Name'].replace(r'.*Mr.|.Mrs.*', 'Adult', regex=True)

df['Name'].head()
````

### Replace NaN with a Value
`df['Cabin'].replace(np.nan, 'Unknown')` — equivalent to `df['Cabin'].fillna('Unknown')`.  
`.fillna()` is generally preferred for missing value replacement (more explicit intent).


````python
#Replace NaN with a Value
import numpy as np
df['Cabin'] = df['Cabin'].replace(np.nan, 'Unknown')
````

### 📝 Practice — Replacing Values


````python
#Practice Activity: Replacing Values
import pandas as pd

df = pd.read_csv('titanic.csv')
````

#### Task 1 — Encode Sex as 0/1
`df['Sex'].replace({'male': 0, 'female': 1})` — numeric encoding for ML.  
Note: `np.where(df['Sex']=='male', 0, 1)` is an alternative.


````python
#Task 1: In the Sex column, replace 'male' with 0 and 'female' with 1. Verify with .value_counts().

df['Sex'] = df['Sex'].replace(['male', 'female'], [0 , 1])
df['Sex'].value_counts()
````

#### Task 2 — Decode Embarked Codes
`df['Embarked'].replace({'S':'Southampton','C':'Cherbourg','Q':'Queenstown'})` — human-readable labels.


````python
"""Task 2: In the Embarked column, replace 'S' with 'Southampton', 'C' with 'Cherbourg', and 'Q' with 'Queenstown'. 
Use a dictionary mapping."""

df['Embarked'] = df['Embarked'].replace({'S': 'Southapton',
                                         'C': 'Cherbourg',
                                         'Q': 'Queenstown'})
df['Embarked'].value_counts()
````

#### Task 3 — Replace 0/1 with 'No'/'Yes'
`df['Survived'].replace({0:'No', 1:'Yes'})` — makes the column self-documenting.


````python
"""Task 3: The Survived column contains 0 and 1. Replace these with 'No' and 'Yes' respectively. 
Display the first 5 rows."""

df['Survived'] = df['Survived'].replace([0, 1], ['No', 'Yes'])

df['Survived'].head()
````

#### Task 4 — Replace Across the Entire DataFrame
`df.replace({'male':'M', 'female':'F'})` — replaces matching values in **every column**.  
Useful for global corrections (e.g., fixing a typo that appears in multiple columns).


````python
"""Task 4: Use a single .replace() call on the entire DataFrame to replace all 
occurrences of 'male' with 'M' and 'female' with 'F'. (Hint: pass a nested dictionary.)"""
df.replace({'Sex': {'male': 'M', 'female': 'F'}}, inplace=True)
````

#### Task 5 (Bonus) — Regex Replace on Names
`df['Name'].replace(r'.*(Miss|Mrs|Ms).*', 'Female_Title', regex=True)` — replaces any name containing a female title.  
The `|` in regex means OR — any of the three patterns triggers the replacement.


````python
"""Task 5 (Bonus): In the Name column, use regex=True to replace any name containing 'Miss' or 'Mrs' or 'Ms' 
with 'Female_Title'. Display the first 10 names."""
df['Name'] = df['Name'].str.replace(r'\b(Miss|Mrs|Ms)\.?\b', 'Female_Title', regex=True)
df['Name'].head(10)
````
