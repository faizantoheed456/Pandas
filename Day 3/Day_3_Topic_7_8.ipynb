```python
#Day 3, Topic 8: Replacing Values

import pandas as pd

df = pd.read_csv('titanic.csv')
```

```python
#Replace a Single Value
df['Sex'] = df['Sex'].replace('male', 'M')
df['Sex'].value_counts()
```

```python
#Replace Multiple Values
df['Embarked'] = df['Embarked'].replace(['C', 'Q'], 'other')
df['Embarked'].value_counts()
```

```python
#Replace Using a Dictionary

embark_mapping = {
    'S': 'Southampton', 
    'C': 'Cherbourg',
    'Q': 'Queenstown'
}

df['Embarked'] = df['Embarked'].replace(embark_mapping)
df['Embarked'].value_counts()
```

```python
#Replace Using Regular Expressions
df['Name'] = df['Name'].replace(r'.*Mr.|.Mrs.*', 'Adult', regex=True)

df['Name'].head()
```

```python
#Replace NaN with a Value
import numpy as np
df['Cabin'] = df['Cabin'].replace(np.nan, 'Unknown')
```

```python
#Practice Activity: Replacing Values
import pandas as pd

df = pd.read_csv('titanic.csv')
```

```python
#Task 1: In the Sex column, replace 'male' with 0 and 'female' with 1. Verify with .value_counts().

df['Sex'] = df['Sex'].replace(['male', 'female'], [0 , 1])
df['Sex'].value_counts()
```

```python
"""Task 2: In the Embarked column, replace 'S' with 'Southampton', 'C' with 'Cherbourg', and 'Q' with 'Queenstown'. 
Use a dictionary mapping."""

df['Embarked'] = df['Embarked'].replace({'S': 'Southapton',
                                         'C': 'Cherbourg',
                                         'Q': 'Queenstown'})
df['Embarked'].value_counts()
```

```python
"""Task 3: The Survived column contains 0 and 1. Replace these with 'No' and 'Yes' respectively. 
Display the first 5 rows."""

df['Survived'] = df['Survived'].replace([0, 1], ['No', 'Yes'])

df['Survived'].head()
```

```python
"""Task 4: Use a single .replace() call on the entire DataFrame to replace all 
occurrences of 'male' with 'M' and 'female' with 'F'. (Hint: pass a nested dictionary.)"""
df.replace({'Sex': {'male': 'M', 'female': 'F'}}, inplace=True)
```

```python
"""Task 5 (Bonus): In the Name column, use regex=True to replace any name containing 'Miss' or 'Mrs' or 'Ms' 
with 'Female_Title'. Display the first 10 names."""
df['Name'] = df['Name'].str.replace(r'\b(Miss|Mrs|Ms)\.?\b', 'Female_Title', regex=True)
df['Name'].head(10)
```
