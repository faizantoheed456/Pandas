```python
#Day 3, Topic 7: Renaming Columns
import pandas as pd
df = pd.read_csv('titanic.csv')
```

```python
#.rename() – The Primary Tool
df = df.rename(columns={
    'Pclass': 'Passenger_Class',
    'SibSp': 'Siblings_Spouses'
})

df.columns.tolist()
```

```python
#Renaming All Columns at Once

df.columns = df.columns.str.lower()
df.columns.tolist()
```

```python
#Replace spaces with underscores:
df.columns.str.replace(' ', '_')
```

```python
#Using a Function to Transform Names
# Convert all columns to uppercase
df = df.rename(columns=str.upper)

# Or use a lambda
df = df.rename(columns=lambda x: x.replace('_', ''))
```

```python
#Adding Prefixes or Suffixes
df_prefixed = df.add_prefix('col_')

df_suffixed = df.add_suffix('_titanic')

```

```python
#Titanic Column Standardization
df.columns = (df.columns.str.lower().str.replace(' ', '_').str.replace('[^a-z0-9_]', '', regex=True))
```

```python
#Practice Activity: Renaming Columns
import pandas as pd

df = pd.read_csv('titanic.csv')
```

```python
#Task 1: Rename the column 'Pclass' to 'Class' and 'SibSp' to 'Siblings'. Verify by printing the column list.

df = df.rename(columns={
    'Pclass': 'Class', 
    'SibSp': 'Siblings'
})

df.columns.tolist()
```

```python
#Task 2: Convert all column names to lowercase. Print the new column list.

df.columns = df.columns.str.lower()

df.columns.tolist()
```

```python
#Task 3: Replace all spaces in column names with underscores. 
# (Titanic has no spaces, so rename one column to 'Passenger Class' first, then fix it.)
df = df.rename(columns={'class': 'Passenger class'})
df.columns = df.columns.str.replace(' ', '_')
df.columns.tolist()
```

```python
#Task 4: Use .rename() with a function to add the suffix '_col' to every column name. (Hint: lambda x: x + '_col')

df = df.rename(columns=lambda x: x + '_col')
df.columns.tolist()
```

```python
"""Task 5 (Bonus): Chain .rename() and .str methods to convert all column names to lowercase, 
replace spaces with underscores, and remove any periods. Apply this to the original Titanic DataFrame."""

df = df.rename(columns=lambda x: x.lower().replace(' ', '_').replace('.', ''))
df.columns
```

```python
#Day 3, Topic 8: Replacing Values

import pandas as pd

df = pd.read_csv('titanic.csv')
```

```python
#Replace a Single Value
df['Sex'] = df['Sex'].replace('male', 'M')
df['Sex'].value_counts()
```

```python
#Replace Multiple Values
df['Embarked'] = df['Embarked'].replace(['C', 'Q'], 'other')
df['Embarked'].value_counts()
```

```python
#Replace Using a Dictionary

embark_mapping = {
    'S': 'Southampton', 
    'C': 'Cherbourg',
    'Q': 'Queenstown'
}

df['Embarked'] = df['Embarked'].replace(embark_mapping)
df['Embarked'].value_counts()
```
