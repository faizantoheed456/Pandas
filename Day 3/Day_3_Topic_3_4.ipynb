```python
#Practice Activity: Filling Missing Data

import pandas as pd

df = pd.read_csv('titanic.csv')
```

```python
"""#Task 1: Fill missing Age values with the median age. 
# Verify that there are no missing Age values remaining."""
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Age'].isna().sum()
```

```python
"""Task 2: Fill missing Embarked values with the mode (most frequent value). 
What is the mode? Verify no missing Embarked remain."""
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
df['Embarked'].isna().sum()
```

```python
"""Task 3: Create a new column Cabin_filled that fills missing Cabin values with the 
string 'Unknown'. Display the value counts of Cabin_filled.
"""
df['Cabin_filled'] = df['Cabin'].fillna('Unknown')
df['Cabin_filled'].value_counts()
```

```python
"""Task 4: Use forward fill on the Age column (create a copy Age_ffill). 
Compare the first 10 rows of original Age, Age_ffill, and median-filled Age. 
What do you notice about forward fill?"""

df['Age_ffill'] = df['Age'].ffill()
df['Age_median'] = df['Age'].fillna(df['Age'].median())
df[['Age', 'Age_ffill', 'Age_median']].head(10)
```

```python
"""Task 5 (Bonus): Fill all missing values in the entire DataFrame at once using a dictionary: 
Age with median, Embarked with mode, Cabin with 'Unknown'. 
Confirm the DataFrame has no missing values."""

fill_values = {
    'Age': df['Age'].median(),
    'Embarked': df['Embarked'].mode(),
    'Cabin': 'Unknown'
}

df_filled = df.fillna(fill_values)

df_filled.isna().sum().sum()
```

```python
#Practice Activity: Handling Duplicates
import pandas as pd
data = {
    'ID': [101, 102, 103, 101, 104, 102, 105],
    'Product': ['A', 'B', 'C', 'A', 'D', 'B', 'E'],
    'Price': [10, 20, 30, 10, 40, 20, 50],
    'Date': ['2024-01-01', '2024-01-02', '2024-01-03', '2024-01-01', '2024-01-04', '2024-01-02', '2024-01-05']
}
df = pd.DataFrame(data)
df
```

```python
#Task 1: Use .duplicated() to identify duplicate rows in the entire DataFrame. 
# How many duplicate rows are there?
dupes = df.duplicated()
dupes.sum()
```

```python
#Task 2: Drop duplicate rows and assign the result to df_unique. Print the cleaned DataFrame.

df_unique = df.drop_duplicates()
df_unique
```

```python
#Task 3: Identify duplicates based only on the ID column (ignore other columns). 
# How many duplicate IDs exist?

id_dupes = df.duplicated(subset=['ID'])
id_dupes.sum()
```

```python
#Task 4: Drop duplicates based on the ID column, but keep the last occurrence of each ID. 
# Print the result.

id_dupes = df.drop_duplicates(subset=['ID'], keep='last')
id_dupes
```

```python
"""Task 5 (Bonus): Find all rows where both Product and Price are duplicated 
(i.e., same product sold at same price on potentially different dates). 
Drop them, keeping the first occurrence.
"""
result = df.drop_duplicates(subset=['Product', 'Price'], keep='first')
result
```
