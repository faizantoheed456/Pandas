```python
#Selection & Filtering
print(f"Selecting the supplier and item type in the 100th row from 'Warehouse and Retail Sales:\n{df.iloc[100, [2, 5]]}")

print(f"Selecting the retail sales, warehouse sales and total sales in the 50 to 60 rows from 'Warehouse and Retail Sales:\n{df.iloc[50:61, [6, 8, 9]]}")

top_supplier = df['SUPPLIER'].value_counts().idxmax()
filtered_rows = df[df['SUPPLIER'] == top_supplier]

print(f"The top supplier from 'Warehouse ad Retail Sales' is '{top_supplier}' with total {len(filtered_rows)} numbers of products.")
```

```python
#Selecting and filtering
moderate_sales = df[(df['TOTAL_SALES'] >= 1000) & (df['TOTAL_SALES'] <= 5000)]
print("Ten rows of total sales which falls between 1000 and 5000: ")
moderate_sales.head(10)
```

```python
#Selecting and filtering
top_suppliers = df['SUPPLIER'].value_counts().index[:2]
top_items = df['ITEM TYPE'].value_counts().index[:2]

filtered_df = df[(df['SUPPLIER'].isin(top_suppliers)) & (df['ITEM TYPE'].isin(top_items))]
print("Five rows of 'Warehouse and Retail Sales' with one of two top suppliers and item types:")
filtered_df.head()
```

```python
#Data cleaning
df = df.drop_duplicates()
print("Rows after removing duplicates:", len(df))
```

```python
#Data cleaning
print(f"Missing count on every column of 'Warehouse and Retail Sales':\n{df.isna().sum()}")
print(f"\nList of columns with missing values:{df.columns[df.isna().any()].tolist()}")
```

```python
#Data cleaning
mask = ((df.isna().sum() / len(df)) * 100) > 0.40
columns_to_drop = mask[mask == True].index
df = df.drop(columns=columns_to_drop)
for col in df.columns:
    if df[col].isna().sum() > 0:
        if df[col].dtype in ['float64', 'int64']:
            df[col] = df[col].fillna(df[col].median())
        else:
            df[col] = df[col].fillna(df[col].mode()[0])
print(f"VERIFICATION; Sum of missing values in each columns:\n{df.isna().sum()}")
```

```python
#Data cleaning
print("Memory before: ", df.memory_usage(deep=True).sum() / 1024**2, "MB")
df['ITEM TYPE'] = df['ITEM TYPE'].astype('category')
df['SUPPLIER'] = df['SUPPLIER'].astype('category')
print("Memory after: ", df.memory_usage(deep=True).sum() / 1024**2, "MB")
```

```python
#Data cleaning
df['SALES_CATEGORY'] = pd.qcut(df['TOTAL_SALES'], q=3, labels=['Low', 'Medium', 'High'])


print("Five rows with a categorical column of 'Warehouse and Retail Sales':")
df[['TOTAL_SALES', 'SALES_CATEGORY']].head(5)
```

```python
df.rename(columns={
    'RETAIL SALES': 'RetailSales',
    'WAREHOUSE SALES': 'WarehouseSales',
    'ITEM TYPE': 'ItemType'
}, inplace=True)

print(f"List of renamed columns of 'Warehouse and Retail Sales':\n{df.columns.tolist()}")
```

```python
#Data cleaning
df['ItemType'].replace({
    'WINE': 'Wine',
    'LIQUOR': 'Liquor',
    'BEER': 'Beer',
    'KEGS': 'Kegs',
    'NON-ALCOHOL': 'Non-Alcohol',
    'STR_SUPPLIES': 'Str_Supplies',
    'REF': 'Ref',
    'DUNNAGE': 'Dunnage'
}, inplace=True)

print(f"Converting all item types into title case: ")
df[['ItemType']].head(10)
```

```python
#Grouping & Aggregation
supplier_stats = df.groupby('SUPPLIER')['RetailSales'].agg(['sum', 'mean', 'count'])
top_5_suppliers = supplier_stats.sort_values(by='sum', ascending=False)
print("Supplier stats:\n")
top_5_suppliers.head()
```

```python
#Grouping and aggregations
item_type = df.groupby('ItemType')['RetailSales'].sum()
print(f"ItemType '{item_type.idxmax()}' with total retail sales of ${item_type.max()}")
```

```python
#Grouping and aggregations
supplier_item_types = df.groupby(['SUPPLIER', 'ItemType'])['RetailSales'].agg('sum')
sorted_values = supplier_item_types.sort_values(ascending=False)
print("Three combination with highest sales: ")
sorted_values.head(3)
```

```python
##Grouping and aggregations
df['AVERAGE RETAIL SALES'] = df.groupby('SUPPLIER')['RetailSales'].transform('mean')
df['DIFFERENCE RETAIL SALES'] = df['RetailSales'] - df['AVERAGE RETAIL SALES']
print("Adding two new columns in 'Warehouse and Retail Sales':")
df[['RetailSales', 'AVERAGE RETAIL SALES', 'DIFFERENCE RETAIL SALES']].head(15)
```

```python
#Grouping and aggregations
filtered_df = df.groupby('SUPPLIER').filter(lambda x: len(x) >= 100)
num_suppliers = filtered_df['SUPPLIER'].nunique()
print(f"Number of suppliers with 100+ transactions: {num_suppliers}")
```

```python
#Grouping and aggregations
summary_table = pd.pivot_table(df, values='RetailSales', index='SUPPLIER', columns='ItemType', fill_value=0, margins=True, margins_name='Total')
print("Summary table:")
summary_table
```

```python
#Grouping and aggregations
freq_table = pd.crosstab(df['SUPPLIER'], df['YEAR'])

print("Frequency of supplier and year tables: ")
freq_table.head()
```

```python
#Split and restack
earliest_year = df['YEAR'].min()
earliest_rows = df[df['YEAR'] == earliest_year]
later_rows = df[df['YEAR'] != earliest_year]

combined = pd.concat([earliest_rows, later_rows], axis=0, ignore_index=True)
print(f"Earliest year rows: {len(earliest_rows)}")
print(f"Later rows: {len(later_rows)}")
print(f"Combined rows: {len(combined)} (original: {len(df)})")
print("Combined matches original:", len(combined) == len(df))
```

```python
#Supplier summary and merge
supplier_summary = df.groupby('SUPPLIER').agg(
    TotalRetailSales=('RetailSales', 'sum'),
    UniqueItemTypes=('ItemType', 'nunique')
).reset_index()

# Merge back to original
df_enriched = df.merge(supplier_summary, on='SUPPLIER', how='left')
print("\nEnriched dataset (first 5 rows):")
df_enriched[['SUPPLIER', 'RetailSales', 'TotalRetailSales', 'UniqueItemTypes']].head()
```

```python
#Simulate supplier regions 
import numpy as np
regions = ['North', 'South', 'East', 'West']
unique_suppliers = df[['SUPPLIER']].drop_duplicates()
unique_suppliers['Region'] = np.random.choice(regions, size=len(unique_suppliers))

df_with_region = df.merge(unique_suppliers, on='SUPPLIER', how='left')
print("\nDataset with supplier region (first 5 rows):")
df_with_region[['SUPPLIER', 'RetailSales', 'Region']].head()
```

```python
#Top 5 and bottom 5 suppliers
transaction_counts = df.groupby('SUPPLIER').size().reset_index(name='TransactionCount')
transaction_counts_sorted = transaction_counts.sort_values('TransactionCount')
top_5 = transaction_counts_sorted.tail(5)
bottom_5 = transaction_counts_sorted.head(5)

top_5_sales = top_5.merge(supplier_summary, on='SUPPLIER')[['SUPPLIER', 'TransactionCount', 'TotalRetailSales']]
bottom_5_sales = bottom_5.merge(supplier_summary, on='SUPPLIER')[['SUPPLIER', 'TransactionCount', 'TotalRetailSales']]

top_and_bottom = pd.concat([top_5_sales, bottom_5_sales], axis=0, ignore_index=True)
print("\nTop and bottom 5 suppliers by transaction count:")
top_and_bottom
```

```python
#Set DATE as index and sort chronologically
ts_df = df.set_index('DATE').sort_index()
ts_df = ts_df[['RetailSales', 'WarehouseSales']]  # keep only numeric columns we care about
print("First 5 rows of time‑series DataFrame:")
ts_df.head()
```

```python
#Total retail sales per calendar quarter
quarterly_sales = ts_df['RetailSales'].resample('QE').sum()
print("\nTotal retail sales per quarter:")
quarterly_sales.head()
```

```python
#3‑month rolling average of monthly retail sales
monthly_sales = ts_df['RetailSales'].resample('ME').sum()
rolling_3m = monthly_sales.rolling(window=3).mean()
print("\nMonthly sales with 3‑month rolling average (first 6 months):")
pd.DataFrame({
    'Monthly': monthly_sales,
    '3M_Rolling_Avg': rolling_3m
}).head(6)
```

```python
#Month‑over‑month change
monthly_with_prev = pd.DataFrame({
    'CurrentMonth': monthly_sales,
    'PreviousMonth': monthly_sales.shift(1)
})
monthly_with_prev['PctChange'] = (
    (monthly_with_prev['CurrentMonth'] - monthly_with_prev['PreviousMonth']) 
    / monthly_with_prev['PreviousMonth'] * 100
).round(2)
print("\nMonth‑over‑month percentage change (first 6 rows):")
monthly_with_prev.head(6)
```

```python
#Time zone conversion demonstration
example_utc = pd.Series(pd.to_datetime(['2024-01-01 12:00:00', '2024-07-01 12:00:00']))
example_utc = example_utc.dt.tz_localize('UTC')
example_eastern = example_utc.dt.tz_convert('US/Eastern')
print("\nTime zone conversion example:")
pd.DataFrame({
    'UTC': example_utc,
    'US_Eastern': example_eastern
})
```

```python
#Pivot table – ItemType (rows) × YEAR (columns) with total RetailSales
pivot_retail = pd.pivot_table(
    df,
    values='RetailSales',
    index='ItemType',
    columns='YEAR',
    aggfunc='sum',
    fill_value=0
)
print("Pivot table (ItemType × YEAR) – total RetailSales:")
pivot_retail
```

```python
#Melt the pivot table back to long format
melted = pivot_retail.reset_index().melt(
    id_vars='ItemType',
    var_name='YEAR',
    value_name='TotalRetailSales'
)
print("\nMelted back to long format (first 5 rows):")
melted.head()

# Verify that the melted total matches the original grouped total
original_total = df.groupby(['ItemType', 'YEAR'])['RetailSales'].sum().reset_index()
# Sort both for comparison
melted_sorted = melted.sort_values(['ItemType', 'YEAR']).reset_index(drop=True)
original_sorted = original_total.sort_values(['ItemType', 'YEAR']).reset_index(drop=True)
print("\nMelted and original totals are equal:", melted_sorted.equals(original_sorted))
```

```python
#Same pivot table via groupby + unstack 
# First build a Series with MultiIndex (YEAR, ItemType)
grouped_series = df.groupby(['YEAR', 'ItemType'])['RetailSales'].sum()
# Unstack YEAR to columns
pivot_via_unstack = grouped_series.unstack(level='YEAR', fill_value=0)
# Optionally reorder columns to match Task 24 (years ascending) and index (alphabetical)
pivot_via_unstack = pivot_via_unstack.sort_index(axis=0).sort_index(axis=1)
print("\nPivot table via groupby + unstack:")
pivot_via_unstack
```

```python
#Task 27: MultiIndex and extraction 
# Find the supplier with the most rows
top_supplier = df['SUPPLIER'].value_counts().idxmax()
print(f"Supplier with most rows: '{top_supplier}'")

# Set a two‑level row index
multi = df.set_index(['SUPPLIER', 'ItemType']).sort_index()
print("\nMultiIndex DataFrame (first 5 rows):")
multi.head()

# Extract all rows for the top supplier using .xs()
top_supplier_rows = multi.xs(key=top_supplier, level='SUPPLIER')
print(f"\nRows for '{top_supplier}':")
top_supplier_rows.head()
```

```python
#Task 28: Group by index level 
# Compute total RetailSales per supplier using the index level
totals_by_supplier = multi.groupby(level='SUPPLIER')['RetailSales'].sum()
print("\nTotal RetailSales per supplier (first 5):")
totals_by_supplier.head()
```

```python
# Check current memory usage
mem_before = df.memory_usage(deep=True).sum()
print(f"Memory before: {mem_before / 1024:.2f} KB")

# Convert object columns with few unique values to 'category'
for col in df.select_dtypes(include='object').columns:
    unique_ratio = df[col].nunique() / len(df)
    if unique_ratio < 0.5:   # less than 50% unique
        df[col] = df[col].astype('category')

# Downcast numeric columns
for col in df.select_dtypes(include=['int64', 'float64']).columns:
    df[col] = pd.to_numeric(df[col], downcast='integer' if df[col].dtype == 'int64' else 'float')

mem_after = df.memory_usage(deep=True).sum()
print(f"Memory after:  {mem_after / 1024:.2f} KB")
print(f"Savings:       {(1 - mem_after / mem_before) * 100:.1f}%")
```

```python
chain_result = (
    df
    .query('RetailSales > 0')
    .assign(TotalSales=lambda x: x['RetailSales'] + x['WarehouseSales'])
    .groupby('ItemType')
    .agg(Total=('TotalSales', 'sum'), Average=('TotalSales', 'mean'))
    .sort_values('Total', ascending=False)
)
print("\nChained aggregation result:")
print(chain_result)

# ===== Task 27: Vectorized string filtering =====
mask = df['ITEM DESCRIPTION'].str.contains('750ML', na=False)
matching_rows = df[mask]
print(f"\nNumber of rows with '750ML' in description: {len(matching_rows)}")
print("First few matching rows:")
matching_rows[['ITEM DESCRIPTION', 'ItemType', 'RetailSales']].head()
```
