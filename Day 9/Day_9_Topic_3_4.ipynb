```python
#Day 9, Topic 3: .query() for Speed

import pandas as pd
import numpy as np
import time
```

```python
#Timing Comparison
n = 500_000
np.random.seed(42)
large_df = pd.DataFrame({
    'Age': np.random.randint(0, 100, n),
    'Sex': np.random.choice(['male', 'female'], n),
    'Fare': np.random.rand(n) * 500
})

# Boolean indexing
start = time.perf_counter()
masked = large_df[(large_df['Age'] > 18) & (large_df['Sex'] == 'male')]
time_bool = time.perf_counter() - start
print(f"Boolean indexing: {time_bool:.4f} s")

# .query()
start = time.perf_counter()
queried = large_df.query("Age > 18 and Sex == 'male'")
time_query = time.perf_counter() - start
print(f".query(): {time_query:.4f} s")

print(f"Speedup: {time_bool / time_query:.2f}x")
```

```python
#Memory Comparison
import sys
temp_bool = (large_df['Age'] > 18) & (large_df['Sex'] == 'male')
print(f"Boolean mask memory: {sys.getsizeof(temp_bool) / 1024:.2f} KB")

```

```python
#When .query() Shines
min_age = 30
result = large_df.query("Age > @min_age")
```

```python
#Practice Activity: .query() Speed
import pandas as pd
import numpy as np
import time
import sys
```

```python
n = 2_000_000
np.random.seed(42)
df = pd.DataFrame({
    'A': np.random.rand(n),
    'B': np.random.rand(n),
    'C': np.random.rand(n)
})

# Task 1: Boolean indexing
start = time.perf_counter()
bool_result = df[(df['A'] > 0.5) & (df['B'] < 0.3)]
time_bool = time.perf_counter() - start
print(f"Boolean indexing: {time_bool:.4f} seconds")
```

```python
# Task 2: .query()
start = time.perf_counter()
query_result = df.query("A > 0.5 and B < 0.3")
time_query = time.perf_counter() - start
print(f".query()       : {time_query:.4f} seconds")
```

```python
# Task 3: Speedup
speedup = time_bool / time_query
print(f"\nSpeedup factor: {speedup:.2f}x")
print(f".query() is {speedup:.1f} times faster than Boolean indexing.")
```

```python
#Day 9, Topic 4: Vectorized String Operations vs. .apply()
import pandas as pd
import numpy as np
import time

df = pd.read_csv('titanic.csv')
```

```python
# .str.len() – vectorized
start = time.perf_counter()
df['name_len_str'] = df['Name'].str.len()
time_str = time.perf_counter() - start
print(f".str.len() : {time_str:.6f} s")

# .apply() with len – per‑row Python loop
start = time.perf_counter()
df['name_len_apply'] = df['Name'].apply(len)
time_apply = time.perf_counter() - start
print(f".apply(len): {time_apply:.6f} s")

print(f"Speedup   : {time_apply / time_str:.1f}x")
```

```python
#Simulating a Larger Dataset
big_names = pd.concat([df['Name']] * 50_000, ignore_index=True)
print(f"Length: {len(big_names):,}")

# .str.len()
start = time.perf_counter()
big_names.str.len()
time_str = time.perf_counter() - start
print(f".str.len() : {time_str:.4f} s")

# .apply(len)
start = time.perf_counter()
big_names.apply(len)
time_apply = time.perf_counter() - start
print(f".apply(len): {time_apply:.4f} s")

print(f"Speedup: {time_apply / time_str:.1f}x")
```

```python
#Practice Activity: Vectorized Strings vs .apply()
import pandas as pd
df = pd.read_csv('titanic.csv')
```

```python
# Task 1: Vectorized .str.split()
start = time.perf_counter()
last_name_str = df['Name'].str.split(',').str[0]
time_str = time.perf_counter() - start
print(f".str.split(): {time_str:.6f} seconds")
```

```python
# Task 2: .apply() with lambda
start = time.perf_counter()
last_name_apply = df['Name'].apply(lambda x: x.split(',')[0])
time_apply = time.perf_counter() - start
print(f".apply():     {time_apply:.6f} seconds")
```

```python
# Task 3: Speedup factor
speedup = time_apply / time_str
print(f"Speedup: {speedup:.1f}x faster using .str.split()")

# Check that results are identical
assert (last_name_str == last_name_apply).all(), "Results differ!"
print("Both methods produce identical results.")
```
