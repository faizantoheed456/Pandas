```python
#Practice Activity: Memory Optimization
import pandas as pd
df = pd.read_csv('titanic.csv')
```

```python
#Task 1: Print the baseline memory usage using info(memory_usage='deep'). Note the total memory used.

baseline_mem = df.memory_usage(deep=True).sum()
baseline_mem
```

```python
"""Task 2: Convert Sex, Embarked, and Pclass to category dtype. Print the new memory usage. 
How much did you save on these columns?"""

# Convert to category
for col in ['Sex', 'Embarked', 'Pclass']:
    df[col] = df[col].astype('category')

print("\n--- Memory Usage After Categorical Conversion ---")
df.info(memory_usage='deep')
category_mem = df.memory_usage(deep=True).sum()

savings = baseline_mem - category_mem
print(f"\nSaved {savings / 1024:.2f} KB on these three columns.")
```

```python
# Downcast integers and floats
for col in ['PassengerId', 'Survived', 'SibSp', 'Parch']:
    df[col] = pd.to_numeric(df[col], downcast='integer')

for col in ['Age', 'Fare']:
    df[col] = pd.to_numeric(df[col], downcast='float')

final_mem = df.memory_usage(deep=True).sum()
reduction_pct = (1 - (final_mem / baseline_mem)) * 100

print(f"\nFinal Memory Usage: {final_mem / 1024:.2f} KB")
print(f"Total Reduction: {reduction_pct:.2f}%")
```

```python
def optimize_dtypes(df):
    start_mem = df.memory_usage(deep=True).sum()
    
    for col in df.columns:
        col_type = df[col].dtype
        
        # 1. Convert objects to category if unique values < 50%
        if col_type == 'object':
            num_unique = df[col].nunique()
            num_total = len(df)
            if (num_unique / num_total) < 0.5:
                df[col] = df[col].astype('category')
                
        # 2. Downcast numeric columns
        elif 'int' in str(col_type):
            df[col] = pd.to_numeric(df[col], downcast='integer')
        elif 'float' in str(col_type):
            df[col] = pd.to_numeric(df[col], downcast='float')
            
    end_mem = df.memory_usage(deep=True).sum()
    print(f"Optimization complete. Saved {(start_mem - end_mem) / 1024:.2f} KB.")
    return df

# Test the function
df_optimized = optimize_dtypes(df.copy())
```

```python
#Day 9, Topic 2: Method Chaining
import pandas as pd
```

```python
#method chaining:
df = (pd.read_csv('titanic.csv')
     .assign(FamilySize= lambda x: x['SibSp'] + x['Parch'] + 1)
     .rename(columns={'Sex': 'Gender'})
     .drop(columns=['Ticket', 'Cabin'])
     .query('Age > 18')
)

df.head()
```

```python
#Using .pipe() for Custom Functions
def add_family_size(df):
    df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
    return df

def categorize_age(df):
    df['AgeGroup'] = pd.cut(df['Age'], bins=[0, 18, 35, 60, 100],
                                       labels=['Child', 'Young Adult', 'Adult', 'Senior'])
    return df


df = (pd.read_csv('titanic.csv')
     .pipe(add_family_size)
     .pipe(categorize_age)
     .query('AgeGroup == "Adult"')
     .head()
)

df[['Name', 'Age', 'AgeGroup']]
```

```python
import numpy as np
df = pd.read_csv('titanic.csv').assign(
    FamilySize = lambda x: x['SibSp'] + x['Parch'] + 1,
    IsAlone = lambda x: x['FamilySize'] == 1,
    FareLog = lambda x: np.log1p(x['Fare'])
)

df[['FamilySize', 'IsAlone', 'FareLog']].head()
```

```python
#Chaining After GroupBy
result = (pd.read_csv('titanic.csv')
         .groupby('Pclass')
         .agg(AvgFare=('Fare', 'mean'), Count=('PassengerId', 'count'))
         .reset_index()
         .rename(columns={'AvgFare': 'Average_Fare'})
         .query('Count > 100')
)

result
```

```python
#Using loc in a Chain
df = pd.read_csv('titanic.csv').pipe(lambda x: x.loc[x['Pclass'] == 1, ['Name', 'Age']]).head()
```
