```python
#Day 9, Topic 5: Iterating Efficiently
import pandas as pd
import numpy as np
import time
```

```python
df = pd.read_csv('titanic.csv')
```

```python
# Using .iterrows() – SLOW
start = time.perf_counter()
df['Name_Length_Iterrows'] = np.nan
for idx, row in df.iterrows():
    df.at[idx, 'Name_Length_Iterrows'] = len(row['Name'])
time_iterrows = time.perf_counter() - start
print(f".iterrows() : {time_iterrows:.4f} seconds")
```

```python
# Using .itertuples() – FASTER
start = time.perf_counter()
name_lengths = []
for row in df.itertuples(index=False):
    name_lengths.append(len(row.Name))
df['Name_Length_Ituples'] = name_lengths
time_ituples = time.perf_counter() - start
print(f".itertuples(): {time_ituples:.4f} seconds")
print(f"Speedup over .iterrows(): {time_iterrows / time_ituples:.1f}x")
```

```python
# Vectorized – FASTEST
start = time.perf_counter()
df['Name_Length_Vec'] = df['Name'].str.len()
time_vec = time.perf_counter() - start
print(f"Vectorized : {time_vec:.4f} seconds")
print(f"Speedup over .iterrows(): {time_iterrows / time_vec:.1f}x")
```

```python
#Practice Activity: Efficient Iteration
import pandas as pd
import numpy as np
import time

df = pd.read_csv('titanic.csv')
```

```python
# Task 1: Using .iterrows() (slowest)
start = time.perf_counter()
df['FamilySize_iter'] = None
for idx, row in df.iterrows():
    df.at[idx, 'FamilySize_iter'] = row['SibSp'] + row['Parch'] + 1
time_iterrows = time.perf_counter() - start
print(f".iterrows()   : {time_iterrows:.4f} seconds")
```

```python
# Task 2: Using .itertuples() (faster)
start = time.perf_counter()
family_sizes = []
for row in df.itertuples(index=False):
    family_sizes.append(row.SibSp + row.Parch + 1)
df['FamilySize_tup'] = family_sizes
time_ituples = time.perf_counter() - start
print(f".itertuples() : {time_ituples:.4f} seconds")
```

```python
# Task 3: Vectorized (fastest)
start = time.perf_counter()
df['FamilySize_vec'] = df['SibSp'] + df['Parch'] + 1
time_vec = time.perf_counter() - start
print(f"Vectorized    : {time_vec:.4f} seconds")
```

```python
# Task 4: Speedup report
print(f"\nSpeedup .itertuples() over .iterrows(): {time_iterrows / time_ituples:.1f}x")
print(f"Speedup vectorized over .iterrows(): {time_iterrows / time_vec:.1f}x")
print(f"\nAll columns equal? {df['FamilySize_iter'].equals(df['FamilySize_vec']) & df['FamilySize_tup'].equals(df['FamilySize_vec'])}")
```
