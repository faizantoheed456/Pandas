```python
#Day 8, Topic 3: Selecting from MultiIndex
import pandas as pd
df = pd.read_csv('titanic.csv')
```

```python
# Create a MultiIndex DataFrame
multi = df.groupby(['Pclass', 'Sex', 'Embarked']).agg(
    Total_Fare=('Fare', 'sum'),
    Passengers=('PassengerId', 'count')
)
multi.head(10)
```

```python
#Selecting by First Level
multi.loc[1]
```

```python
#Selecting by a Tuple
multi.loc[(1, 'male')]
```

```python
#Slicing on MultiIndex
multi_sorted = multi.sort_index()

multi_sorted.loc[(1, 'male'):(2, 'female')]
```

```python
#Using slice(None) and IndexSlice
idx = pd.IndexSlice

result = multi_sorted.loc[idx[1, :, 'C'], :]
result
```

```python
#Selecting Columns with MultiIndex Rows
multi.loc[1, 'Total_Fare']
```

```python
#Practice Activity: Selecting from MultiIndex
import pandas as pd

df = pd.read_csv('titanic.csv')
# Create a MultiIndex DataFrame
multi = df.groupby(['Pclass', 'Sex', 'Embarked']).agg(
    Total_Fare=('Fare', 'sum'),
    Passengers=('PassengerId', 'count')
)
multi.head(10)
```

```python
#Slicing on MultiIndex
multi_sorted = multi.sort_index()
multi_sorted.head(10)
```

```python
#Task 1: Select all rows where Pclass=2 and Sex='female'. Display the result.

multi_sorted.loc[(2, 'female')]
```

```python
#Task 2: Using the sorted MultiIndex (multi_sorted), slice from (1, 'male', 'C') to (2, 'female', 'S') inclusive. Show the resulting rows.

range_slice = multi_sorted.loc[(1, 'male', 'C') : (2, 'female', 'S')]
range_slice
```

```python
#Task 3: Use IndexSlice to select all rows where Pclass=3 and Embarked='S', for any sex. Display the result.

idx = pd.IndexSlice

multi_sorted.loc[idx[3, :, 'S'], :]
```

```python
#Task 4 (Bonus): Using .loc[] with a tuple, select only the 'Passengers' column for (2, 'male', 'C'). Print the value.

multi_sorted.loc[(2, 'male', 'C'), 'Passengers']
```

```python
#Day 8, Topic 4: Cross‑Section with .xs()
import pandas as pd
df = pd.read_csv('titanic.csv')

multi = df.groupby(['Pclass', 'Sex', 'Embarked']).agg(
    Total_Fare=('Fare', 'sum'),
    Passengers=('PassengerId', 'count')
)
multi_sorted = multi.sort_index()
multi_sorted.head(6)
```

```python
#select all rows where Sex='male', regardless of Pclass or Embarked.

males = multi_sorted.xs('male', level='Sex')
males
```

```python
#use drop_level=False
males_with_level = multi_sorted.xs('male', level='Sex', drop_level=False)
males_with_level
```

```python
#Selecting by Level Number Instead of Name
embarked_C = multi_sorted.xs('C', level=2)
embarked_C.head()
```

```python
#Using .xs() with axis=1 (Columns)
cols = pd.MultiIndex.from_product([['Fare', 'Age'], ['mean', 'median']])
demo = pd.DataFrame([[1, 2, 3, 4], [5, 6, 7, 8]], columns=cols)

demo.xs('mean', level=1, axis=1)
```

```python
#Practice Activity: .xs() on Titanic MultiIndex
import pandas as pd
df = pd.read_csv('titanic.csv')

multi = df.groupby(['Pclass', 'Sex', 'Embarked']).agg(
    Total_Fare=('Fare', 'sum'),
    Passengers=('PassengerId', 'count')
)
multi_sorted = multi.sort_index()
multi_sorted.head(6)
```

```python
#Task 1: Use .xs() to select all rows where Embarked='S'. Display the first 3 rows of the result.

multi_sorted.xs('S', level='Embarked')
```

```python
#Task 2: Select all rows for Pclass=2 using .xs(). Display the result.

multi_sorted.xs(2)
```

```python
"""Task 3: Using .xs() with drop_level=False, select Sex='female' while keeping the Sex level in 
the resulting DataFrame. Verify that the Sex column still appears in the index by printing the first 3 rows."""

multi_sorted.xs('female', level='Sex', drop_level=False).head(3)
```

```python
#Task 4 (Bonus): Use .xs() on the columns of the following DataFrame:
cols = pd.MultiIndex.from_product([['Fare', 'Age'], ['mean', 'std']])
demo = pd.DataFrame([[30.5, 12.0, 22.0, 8.5],
                      [70.2, 15.4, 28.0, 6.2]],
                    index=['Male', 'Female'],
                    columns=cols)
demo
```

```python
#Extract only the 'std' statistics using .xs() with level=1 and axis=1. Display the result.

demo.xs('std', level=1, axis=1)
```
