```python
#Day 8, Topic 2: Creating MultiIndex
import pandas as pd
```

```python
#Creating a MultiIndex from Tuples or Arrays
tuples = [('North', 2022), ('North', 2023), ('South', 2022), ('South', 2023)]
multi_index = pd.MultiIndex.from_tuples(tuples, name=['Region', 'Year'])

data = [100, 150, 200, 250]
df_tuples = pd.DataFrame({'Sales': data}, index=multi_index)
df_tuples
```

```python
regions = ['North', 'North', 'South', 'South']
years = [2022, 2023, 2022, 2023]
multi_index2 = pd.MultiIndex.from_arrays([regions, years], names=['Region', 'Year'])
df_arrays = pd.DataFrame({'Sales': data}, index=multi_index2)
df_arrays
```

```python
#Creating a MultiIndex from a Product (Cartesian Product)
regions = ['North', 'South']
years = [2022, 2023]
multi_product = pd.MultiIndex.from_product([regions, years], names=['Region', 'Year'])
df_product = pd.DataFrame(index=multi_product)
df_product['Sales'] = [100, 150, 170, 200]
df_product
```

```python
#Practice Activity: Creating MultiIndex
import pandas as pd
df = pd.read_csv('titanic.csv')
```

```python
"""Task 1: Create a MultiIndex from tuples: (1, 'male'), (1, 'female'), (2, 'male'), (2, 'female'), (3, 'male'), 
(3, 'female'). Assign it to a DataFrame with a column 'Count' 
containing the number of passengers in each group 
(you'll need to compute those counts first). Display the result."""

tuples = [(1, 'male'), (1, 'female'), (2, 'male'), (2, 'female'), (3, 'male'), (3, 'female')]
multi_index = pd.MultiIndex.from_tuples(tuples, names=['Pclass', 'Sex'])

counts = df.groupby(['Pclass', 'Sex']).size()

multi_df = pd.DataFrame(counts, index=multi_index, columns=['Count'])

multi_df
```

```python
"""Task 2: Use pd.MultiIndex.from_product() to create an index of all combinations 
of ['C', 'Q', 'S'] (Embarked) and [0, 1] (Survived). Create an empty DataFrame 
with this index and then fill a column 'Total_Fare' with the sum of Fare for 
each combination (compute the required values). Display the filled DataFrame."""

embarked = ['C', 'Q', 'S']
survived = [0, 1]
multi = pd.MultiIndex.from_product([embarked, survived], names=['Embarked', 'Survived'])

data = pd.DataFrame(index=multi)
data['TotalFare'] = df.groupby(['Embarked', 'Survived'])['Fare'].sum()

data
```

```python
"""Task 3: Set a MultiIndex on the original df using the columns 'Embarked' and 'Pclass'. 
Then use .reset_index(level='Embarked') to move only the 'Embarked' level 
back to columns while keeping Pclass as the index."""

df_set = df.set_index(['Embarked', 'Pclass'])
df_reset = df_set.reset_index(level='Embarked')
df_reset.head()
```

```python
"""Task 4 (Bonus): Using the MultiIndex from Task 1, use .loc[] to select the row for Pclass=1
and Sex='female'. Print the count."""


tuples = [(1, 'male'), (1, 'female'), (2, 'male'), (2, 'female'), (3, 'male'), (3, 'female')]
multi_index = pd.MultiIndex.from_tuples(tuples, names=['Pclass', 'Sex'])

counts = df.groupby(['Pclass', 'Sex']).size()

multi_df = pd.DataFrame(counts, index=multi_index, columns=['Count'])

multi_df.loc[1, 'female']
```
