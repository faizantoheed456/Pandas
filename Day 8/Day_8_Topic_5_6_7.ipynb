```python
#Day 8, Topic 5: Stacking and Unstacking MultiIndex DataFrames
import pandas as pd

df = pd.read_csv('titanic.csv')

# Create a pivot table with MultiIndex columns
pivot = pd.pivot_table(df, values='Fare', index='Pclass', columns=['Sex', 'Embarked'], aggfunc='mean')
print("Original pivot (MultiIndex columns):")
pivot
```

```python
stacked = pivot.stack(level='Sex')
stacked
```

```python
unstacked = stacked.unstack(level='Sex')

unstacked
```

```python
#Day 8, Topic 6: MultiIndex Aggregations (.groupby(level=))
import pandas as pd
```

```python
#Basic .groupby(level=...)
multi = df.groupby(['Pclass', 'Sex', 'Embarked']).agg(
    Total_Fare=('Fare', 'sum'),
    Passengers=('PassengerId', 'count')
)

multi_sorted = multi.sort_index()

by_class = multi_sorted.groupby(level='Pclass').sum()
by_class
```

```python
#Group by multiple index levels:
by_class_sex = multi_sorted.groupby(level=['Pclass', 'Sex']).sum()
by_class_sex
```

```python
#Aggregating with Different Functions
stats = multi_sorted.groupby(level='Pclass').agg({'Total_Fare': 'mean', 'Passengers': 'sum'})
stats
```

```python
#Using Level Numbers
by_sex = multi_sorted.groupby(level=1).sum()
by_sex
```

```python
#Task 1: Group by the 'Embarked' level and compute the mean Total_Fare and sum of Passengers.


multi_sorted.groupby(level='Embarked').agg({'Total_Fare': 'mean', 'Passengers': 'sum'})
```

```python
#Task 2: Group by both 'Pclass' and 'Sex' levels (using level names) and compute the sum of all columns.

multi_sorted.groupby(level=['Pclass', 'Sex']).sum()
```

```python
#Task 3: Group by level 0 (Pclass) and compute the maximum Passengers count.

multi_sorted.groupby(level=0).agg({'Passengers': 'max'})
```

```python
#Task 4 (Bonus): Chain .groupby(level='Sex') and .agg() to compute the median Total_Fare and the mean Passengers per sex.

multi_sorted.groupby(level='Sex').agg({
    'Total_Fare': 'median',
    'Passengers': 'mean'
})
```

```python
#Day 8, Topic 7: Sorting MultiIndex
import pandas as pd
df = pd.read_csv('titanic.csv')

# Create MultiIndex DataFrame (unsorted – the groupby result is not sorted by default)
multi = df.groupby(['Pclass', 'Sex', 'Embarked']).agg(
    Total_Fare=('Fare', 'sum'),
    Passengers=('PassengerId', 'count')
)
print("Unsorted MultiIndex DataFrame:")
multi
```

```python
multi_sorted = multi.sort_index()
multi_sorted
```

```python
#Sorting by Specific Level
multi_sorted_sex = multi.sort_index(level='Sex')
multi_sorted_sex
```

```python
#Descending Order
multi_desc = multi.sort_index(ascending=False)
multi_desc
```
