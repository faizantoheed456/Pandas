```python
#Gauntlet #8: The MultiIndex Master Challenge
import pandas as pd
import numpy as np

df = pd.read_csv('titanic.csv')
```

```python
# Task 1: Create a MultiIndex DataFrame
multi = df.groupby(['Pclass', 'Sex', 'Embarked']).agg(
    Total_Fare=('Fare', 'sum'),
    Avg_Age=('Age', 'mean'),
    Count=('PassengerId', 'count')
)
print("=== Task 1: MultiIndex DataFrame ===")
print(multi.head(10))
```

```python

# Task 2: Sort and Slice

multi_sorted = multi.sort_index()
print("\n=== Task 2: Slice from (1,'male','C') to (2,'female','S') ===")
sliced = multi_sorted.loc[(1, 'male', 'C'):(2, 'female', 'S')]
print(sliced)
```

```python
# Task 3: Cross‑section with .xs() (drop_level=False)
print("\n=== Task 3: All rows for Sex='male' (keeping level) ===")
males = multi_sorted.xs('male', level='Sex', drop_level=False)
print(males.head(3))
```

```python
# Task 4: Aggregation by level (Pclass)
print("\n=== Task 4: Sum of Total_Fare and Count by Pclass ===")
by_class = multi_sorted.groupby(level='Pclass').agg({'Total_Fare': 'sum', 'Count': 'sum'})
print(by_class)
```

```python
# Task 5: Stacking/Unstacking
pivot = pd.pivot_table(df, values='Fare', index='Pclass', columns=['Sex', 'Embarked'], aggfunc='mean')
print("\n=== Task 5: Original pivot table ===")
print(pivot)

stacked = pivot.stack(level='Sex')
print("\nAfter stacking Sex level:")
print(stacked)
```

```python

# Task 6: Reset Index and Re‑create MultiIndex
multi_reset = multi.reset_index()
multi_new = multi_reset.set_index(['Embarked', 'Pclass']).sort_index()
print("\n=== Task 6: New MultiIndex sorted by Embarked, Pclass (first 5 rows) ===")
print(multi_new.head(5))
```

```python
# Task 7 (Bonus): Average Avg_Age for Pclass=1, Sex='female'
avg_age = multi_sorted.loc[(1, 'female')]['Avg_Age']
print("\n=== Task 7 (Bonus): Average Age for Pclass=1, Sex='female' ===")
print(avg_age)
```
