```python
#Day 2, Topic 1: Label-Based Selection with .loc

import pandas as pd
df = pd.read_csv('titanic.csv')
```

```python
#Selecting Rows by Index Label

row_0 = df.loc[0]
print(row_0)
```

```python
#Select multiple rows by passing a list of labels:
row_0_2_5 = df.loc[[0, 2, 5]]
print(row_0_2_5[['Name', 'Age']])
```

```python
#Selecting Rows and Specific Columns
name_0 = df.loc[0, 'Name']
print(name_0)

data_0_2_5 = df.loc[[0, 2, 5], ['Name', 'Age']]
print(data_0_2_5)
```

```python
#Slicing with Labels (Inclusive!)
row_0_to_5 = df.loc[0:5, ['Name', 'Age']]
print(row_0_to_5)
```

```python
#Slice columns by name:
df.loc[:, 'Name':'Age'].head()
```

```python
#Selecting All Rows or All Columns

col_fare = df.loc[:, 'Fare']

row_data = df.loc[0:10, :]
```

```python
#Boolean Indexing with .loc
mask = df['Sex'] == 'female'

women = df.loc[mask, ['Name', 'Age', 'Fare']].head()
print(women)
```

```python
#Combine multiple conditions:

condition = (df['Sex'] == 'female') & (df['Pclass'] == 1) & (df['Age'] > 30)
first_class_older_women = df.loc[condition, ['Name', 'Age', 'Fare']].head()
print(first_class_older_women)
```

```python
#Setting a Meaningful Index

df_indexed = df.set_index('PassengerId')

passenger_5 = df_indexed.loc[5]

print(passenger_5[['Name', 'Pclass']])
```

Practice Activity: .loc

```python
import pandas as pd

df = pd.read_csv('titanic.csv')
```

```python
#Task 1: Select the row with index label 10. Print the passenger's name and age.
row_10 = df.loc[10, ['Name', 'Age']]
print(row_10)
```

```python
#Task 2: Select rows with labels 5, 20, and 50. Display only the Name and Fare columns.
data = df.loc[[5, 20, 50], ['Name', 'Fare']]
print(data)
```

```python
#Task 3: Using label slicing, select rows from label 0 to label 3 (inclusive) and 
# columns from 'Name' to 'Age' (inclusive). Print the result.

data_1 = df.loc[0:3, 'Name':'Age']
print(data_1)
```

```python
#Task 4: Create a Boolean mask for male passengers who survived. Use .loc to display their Name, Age, and Pclass.

mask = df['Sex'] == 'male'

data_2 = df.loc[mask, ['Name', 'Age', 'Pclass']].head()
print(data_2)
```

```python
#Task 5: Set PassengerId as the index. Then use .loc to select the passenger with ID 100. Print their name and fare.

df_indexed = df.set_index('PassengerId')
passenger_100 = df_indexed.loc[100]
print(f"Name: {passenger_100['Name']}, Fare: {passenger_100['Fare']}")
```

```python
"""Task 6 (Bonus): Find all passengers whose name contains 'Mrs'. Use .loc with a 
string condition (.str.contains()) to select their Name and Age.
"""

mrs_passenger = df.loc[df['Name'].str.contains('Mrs', na=False), ['Name', 'Age']]
print(mrs_passenger)
```

Day 2, Topic 2: Position-Based Selection with .iloc

```python
import pandas as pd
 
df = pd.read_csv('titanic.csv')
```

```python
#Selecting Rows by Position
first_row = df.iloc[0]
print(first_row[['Name', 'Age']])
```

```python
#Multiple rows by list:
row_0_2_5 = df.iloc[[0, 2, 5]]
print(row_0_2_5[['Name', 'Age']])
```

```python
#Slicing rows (exclusive end):
first_five = df.iloc[0:5]
print(first_five[['Name', 'Age']])
```

```python
#Selecting Rows and Columns Simultaneously
name_0 = df.iloc[0, 3]
print(name_0)

subset = df.iloc[[0, 2, 5], [3, 5]]
print(subset) 
```

```python
#Slicing Rows and Columns
df.iloc[0:5, 3:6]
```

```python
#Selecting All Rows or All Columns
df.iloc[:, 0:3]

df.iloc[0:10, :]

df.iloc[:, :]
```

```python
#Negative Indexing
df.iloc[-1]

df.iloc[-5:, 0:3]

df.iloc[:-1, -1]
```

```python
#Finding Column Positions
name_pos = df.columns.get_loc('Name')
print(name_pos)

df.iloc[0, name_pos]
```

```python
#Practice Activity: .iloc
import pandas as pd
df = pd.read_csv('titanic.csv')
```

```python
#Task 1: Select the first row using .iloc. Print the passenger's name and class (Pclass).

passenger_info = df.iloc[0]
print(passenger_info[['Name', 'Pclass']])
```

```python
#Task 2: Select the last 3 rows. Display only the Name and Survived columns

df.iloc[-3:, [1, 3]]
```

```python
"""Task 3: Using slicing, select rows at positions 10 through 14 (inclusive of 10, exclusive of 15) 
and columns at positions 2 through 5 (inclusive of 2, exclusive of 6). 
What columns are these? Print the result."""
df.iloc[10:15, 2:6]
```

```python
#Task 4: Select the 5th row (position 4) and the last column (position -1). Print the value.
df.iloc[4, -1]
```

```python
"""Task 5: Set PassengerId as the index. Then use .iloc to select the 100th row (position 99). 
Print the passenger's name.
"""
df_indexed = df.set_index('PassengerId')
passenger_100 = df_indexed.iloc[99]
print(passenger_100['Name'])
```

```python
"""Task 6 (Bonus): Use .iloc to select every 10th row (positions 0, 10, 20, ...) and the columns Name, Age, Fare."""

df.iloc[::10, [3, 5, 9]]
```
