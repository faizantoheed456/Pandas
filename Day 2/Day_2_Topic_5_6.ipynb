Day 2, Topic 5: The .query() Method

```python
import pandas as pd
df = pd.read_csv('titanic.csv')
```

```python
#Basic Syntax
older = df.query("Age > 30")
older[['Name', 'Age']].head()
```

```python
#Multiple Conditions
first_class_womean = df.query("Sex == 'female' and Pclass == 1")
first_class_womean[['Name', 'Sex', 'Pclass']].head()
```

```python
#Using Variables in Queries
min_age = 50
max_fare = 100 
result = df.query("Age > @min_age and Fare < @max_fare")
result[['Name', 'Age', 'Fare']].head()
```

```python
#Practice Activity: .query() Method
import pandas as pd
df = pd.read_csv('titanic.csv')
```

```python
#Task 1: Use .query() to filter passengers older than 60. Display Name and Age.

older = df.query("Age > 60")
older[['Name', 'Age']].head()
```

```python
#Task 2: Use .query() to filter female passengers who survived. How many are there? (Use .shape[0] or len())
survived_womens = df.query("Sex == 'female' and Survived == 1")
len(survived_womens)
```

```python
#Task 3: Use .query() with a variable. Create a variable min_fare = 200. Then filter passengers who paid
# a fare greater than min_fare. Display Name and Fare.

min_fare = 200
fare = df.query("Fare > @min_fare")
fare[['Name', 'Fare']].head()
```

```python
#Task 4: Use .query() to find passengers in first or second class (Pclass in [1, 2]). Display Name and Pclass.
passen = df.query("Pclass in [1, 2]")
passen[['Name', 'Pclass']].head()
```

```python
"""Task 5 (Bonus): Use .query() to find passengers whose name contains 'Mr.'
and who did not survive. Display Name and Survived."""

passen_2 = df[df['Name'].str.contains('Mr.')].query("Survived == 0")
passen_2[['Name', 'Survived']].head() 
```

```python
#Day 2, Topic 6: .isin() and .between()

import pandas as pd

df = pd.read_csv('titanic.csv')
```

```python
#.isin() – Filter by Membership in a List

mask = df['Embarked'].isin(['C', 'Q'])
cherbourg_queenstown = df[mask]
cherbourg_queenstown[['Name', 'Embarked']].head()
```

```python
#.isin() with Multiple Values
upper_class = df[df['Pclass'].isin([1, 2])]

specific_ids = df[df['PassengerId'].isin([10, 20, 30, 40, 50])]
```

```python
#Negating .isin() with ~
# Passengers NOT from Southampton

not_southampton = df[~df['Embarked'].isin(['S'])]
```

```python
#.between() – Filter Numeric Ranges
# Passengers aged 20 to 30 (inclusive)

young_adults = df[df['Age'].between(10, 20)]
young_adults[['Name', 'Age']].head()
```

```python
#Combining .isin() and .between() with Other Conditions
mask = df['Pclass'].isin([1, 2]) & df['Age'].between(30, 50)
result = df[mask]
result[['Name', 'Age', 'Pclass']].head()
```

```python
#Practice Activity: .isin() and .between()
import pandas as pd
df = pd.read_csv('titanic.csv')
```

```python
"""Task 1: Use .isin() to filter passengers who embarked at 'C' (Cherbourg) or 'Q' (Queenstown). 
Display the Name and Embarked columns. How many such passengers are there?"""
mask = df['Embarked'].isin(['C', 'Q'])
result = df[mask]
result[['Name', 'Embarked']].head()
```

```python
#Task 2: Use ~ with .isin() to filter passengers who are not in first or second class. Display Name and Pclass.

mask = ~df['Pclass'].isin([1, 2])
result = df[mask]
result[['Name', 'Pclass']].head()
```

```python
#Task 3: Use .between() to filter passengers aged 18 to 25 (inclusive). Display Name and Age. 
# How many young adults are there?
mask = df['Age'].between(18, 25)
result = df[mask]
result[['Name', 'Age']].head()
len(result)
```

```python
#Task 4: Use .between() with inclusive='left' to filter passengers aged 18 or older but under 30. 
#(i.e., age >= 18 and < 30). Display Name and Age.

mask = df['Age'].between(18, 30, inclusive='left')
result = df[mask]
result[['Name', 'Age']].head()
```

```python
"""Task 5: Combine .isin() and .between(): Find female passengers in first or second class with age between 25 and 40.
Display Name, Sex, Pclass, and Age."""

mask = (df['Sex'] == 'female') & (df['Pclass'].isin([1, 2])) & (df['Age'].between(25, 40))
result = df[mask]
result[['Name', 'Sex', 'Pclass', 'Age']].head()
```
