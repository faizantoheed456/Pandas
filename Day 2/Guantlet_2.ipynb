```python
import pandas as pd
df = pd.read_csv('titanic.csv')
```

```python
"""Label Selection with .loc
Set PassengerId as the index. Then use .loc to select the passenger with ID 200. Print their Name and Age."""

indexed = df.set_index('PassengerId')
passenger_200 = indexed.loc[200]
passenger_200[['Name', 'Age']]
```

```python
"""Position Selection with .iloc
Using the indexed DataFrame from Task 1, select the 50th row (position 49) with .iloc. Print the Name and Fare."""

row_50 = indexed.iloc[49]
row_50[['Name', 'Fare']]
```

```python
"""Boolean Indexing
Filter the DataFrame to show passengers who are female, survived, and are older than 30. 
Display Name, Age, and Fare. How many such passengers?"""

mask = (df['Sex'] == 'female') & (df['Survived'] == 1) & (df['Age'] > 30)
result = df[mask]
result[['Name', 'Age', 'Fare']].head()
print(f"Passengers that are female, survived, and are older than 30: {len(result)}")
```

```python
""".query() Method
Use .query() to find all passengers in first class with fare greater than 100. Display Name, Pclass, and Fare."""

first_class = df.query("Pclass == 1 and Fare > 100")
first_class[['Name', 'Pclass', 'Fare']].head()
```

```python
""".isin() and .between()
Filter passengers whose age is between 20 and 40 (inclusive) AND 
who embarked at 'C' or 'Q'. Display Name, Age, and Embarked."""

mask = (df['Embarked'].isin(['C', 'Q'])) & (df['Age'].between(20, 40))
result = df[mask]
result[['Name', 'Age', 'Embarked']].head(20)
```

```python
df['Title'] = df['Name'].str.extract(r'([A-Za-z]+)\.')
master_passengers = df[df['Title'] == 'Master']
master_passengers[['Name', 'Age']].head()
```

```python
""".where() or .mask()
Use .where() to create a Series high_fare that contains the Fare 
if it's greater than 50, and NaN otherwise. Display the first 10 rows."""

high_fare = df['Fare'].where(df['Fare'] > 50)
high_fare.head(10)
```

```python
""".filter()
Use .filter() to select all columns that end with the letter 'd'. What columns are returned?"""

result = df.filter(regex='d$')
result.columns.tolist()
```
