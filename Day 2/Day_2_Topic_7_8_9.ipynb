Day 2, Topic 7: String Methods (.str Accessor)

```python
import pandas as pd
df = pd.read_csv('titanic.csv')
```

```python
#Filtering with .str.contains()
# Passengers with 'Mrs' in their name
mrs_mask = df['Name'].str.contains('Mrs', case=False)
mrs_passengers = df[mrs_mask]
mrs_passengers[['Name']].head()
```

```python
#Extracting Substrings with .str.extract()
df['Title'] = df['Name'].str.extract(r'([A-Za-z]+)\.')
df[['Name', 'Title']].head()
```

```python
#Splitting Strings with .str.split()
split_names = df['Name'].str.split(',', expand=True)
df['LastName'] = split_names[0]
df['FirstName'] = split_names[1].str.strip()
df[['Name', 'LastName', 'FirstName']].head()

# Get length of last name
df['LastNameLength'] = df['LastName'].str.len()

# Clean and lowercase
df['CleanTitle'] = df['Title'].str.strip().str.lower()
```

```python
#Practice Activity: String Methods

import pandas as pd

df = pd.read_csv('titanic.csv')
```

```python
"""Task 1: Filter the DataFrame to show only passengers whose name contains 'Miss'. 
Display the Name column. How many are there?
"""

miss_mask = df['Name'].str.contains('Miss')
result = df[miss_mask]
result[['Name']].head()
len(result)
```

```python
#Task 2: Filter passengers whose name starts with 'A'. Display Name. How many?

name_mask = df['Name'].str.startswith('A')
result = df[name_mask]
result[['Name']].head()
len(result)
```

```python
"""Task 3: Create a new column NameLength that contains the number of characters in 
each passenger's full name (including spaces). What is the average name length?"""

df['NameLength'] = df['Name'].str.len()
df[['Name', 'NameLength']].head()
avg_name_length = df['NameLength'].mean()
avg_name_length
```

```python
"""Task 4: Extract the title (e.g., 'Mr', 'Mrs', 'Miss', 'Master') from the Name column into a new column called Title. 
Use .str.extract() with the pattern r'([A-Za-z]+)\.'. Display the value counts of Title."""


df['Title'] = df['Name'].str.extract(r'([A-Za-z]+)\.')
df[['Name', 'Title']].head()
df['Title'].value_counts()
```

```python
"""Task 5 (Bonus): Filter passengers whose title is 'Mr' AND who did not survive. Display Name and Survived. 
How many such passengers are there?"""

mr_not_survived = df[(df['Title'] == 'Mr') & (df['Survived'] == 0)]
mr_not_survived[['Name', 'Survived']].head()
len(mr_not_survived)
```

```python
#Day 2, Topic 8: .where() and .mask()

import pandas as pd

df = pd.read_csv('titanic.csv')
```

```python
#.where() – Keep Values Where Condition is True

adult_ages = df['Age'].where(df['Age'] > 18, other='underage')
adult_ages.head(10)
```

```python
# Replace ages ≤ 18 with 'Child'
age_status = df['Age'].mask(df['Age'] <= 18, 'Child')
print(age_status.head(10))
```

```python
# Keep only numeric columns, replace non-numeric with NaN
numeric_df = df.select_dtypes(include='number')
masked_df = numeric_df.where(numeric_df > 0)
print(masked_df.head())
```

```python
# Practice Activity: .where() and .mask()

import pandas as pd

df = pd.read_csv('titanic.csv')
```

```python
"""Task 1: Use .where() to create a Series adult_fares that contains the Fare for passengers aged 18 or older, 
and NaN for those under 18. Display the first 10 rows."""

print("=== Task 1: adult_fares ===")
adult_fares = df['Fare'].where(df['Age'] >= 18)
print(adult_fares.head(10))
```

```python
"""Task 2: Use .mask() to replace all fares greater than 100 with the string 'Expensive'. 
Keep other fares as their original values. Display the first 10 rows."""


result = df['Fare'].mask(df['Fare'] > 100, "Expensive")
result.head(30)
```

```python
"""Task 3: Use .where() on the Sex column to keep only 'female' values and replace 'male' with NaN. 
Display the value counts of the result."""

result = df['Sex'].where(df['Sex'] == 'female')
result.value_counts(dropna=False)
```

```python
#Task 4: Use .mask() to replace all ages less than 18 with 'Minor'. Show the first 10 rows.
result = df['Age'].mask(df['Age'] < 18, 'Minor')
result.head(10)
```

```python
"""Task 5 (Bonus): Create a new column SurvivedText that contains 'Survived' where Survived == 1 and 
'Perished' where Survived == 0. Use either .where() or .mask()"""

df['SurvivedText'] = df['Survived'].mask(df['Survived'] == 1, 'Survived').mask(df['Survived'] == 0, 'Perished')
print(df[['Survived', 'SurvivedText']].head(10))
```

```python
#Day 2, Topic 9: Filtering with .filter()
import pandas as pd

df = pd.read_csv('titanic.csv')
```

```python
#Selecting Columns by Name Pattern
name_columns = df.filter(like='Name')
name_columns.head()
```

```python
#Using regex for Powerful Pattern Matching
ps_columns = df.filter(regex='^(P|S)')
ps_columns.columns.to_list()
```

```python
#Using items to Select Exact Column Names
selected = df.filter(items=['Name', 'Age', 'Fare'])
selected.head()
```
