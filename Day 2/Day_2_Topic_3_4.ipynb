```python
import pandas as pd
df = pd.read_csv('titanic.csv')
```

```python
#Single Bracket – df['column']
age = df['Age']
print(type(age))
print(age.head())
```

```python
#Dot Notation – df.column
name = df.Name
print(name.head())

fare = df.Fare
print(fare.mean())
```

```python
#Double Bracket – df[['col1', 'col2']]
subset = df[['Name', 'Age', 'Fare']]
print(type(subset))
print(subset.head())
```

```python
# Create a new column
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1

# Update an existing column
df['Fare'] = df['Fare'] * 1.1   # 10% increase

# Delete a column (using del)
del df['Ticket']
# Or using drop (safer, returns new DataFrame)
df = df.drop('Cabin', axis=1)
```

```python
#Dynamic Column Selection

selected = [col for col in df.columns if 'Name' in col or 'Age' in col]
df[selected].head()
```

```python
#Day 2, Topic 4: Boolean Indexing (Masking)

import pandas as pd

df =pd.read_csv('titanic.csv')
```

```python
#Part 1: Creating a Boolean Mask
mask = df['Age'] > 30
mask.head(10)
```

```python
#Applying the Mask to Filter Rows
older_passengers = df[mask]
older_passengers[['Name', 'Age']].head()
```

```python
#Combining Multiple Conditions
# Female passengers in first class
mask = (df['Sex'] == 'female') & (df['Pclass'] == 1)
first_class_women = df[mask]
first_class_women[['Name', 'Sex', 'Pclass']].head()
```

```python
# Passengers from first or second class
mask = (df['Pclass'] == 1) | (df['Pclass'] == 2)
first_or_second_class = df[mask]
first_or_second_class[['Name', 'Sex', 'Pclass']].head()
```

```python
#NOT condition: 
# Passengers not from Southampton

mask = ~(df['Embarked'] == 'S')

not_southHampton = df[mask]
```

```python
#Filtering with .loc and Boolean Masks

# Names and fares of female survivors

female_survivors = df.loc[(df['Sex'] == 'female') & (df['Survived'] == 1), ['Name', 'Fare']]
female_survivors
```

```python
#Handling NaN in Boolean Masks
df[df['Age'].isna()].head()
```

```python
#Boolean Masks with String Methods
mrs_mask = df['Name'].str.contains('Mrs')
df[mrs_mask].head()
```

```python
#Practice Activity: Boolean Indexing 
import pandas as pd

df = pd.read_csv('titanic.csv')
```

```python
#Task 1: Filter the DataFrame to show only passengers who survived. How many survivors are there?


(df['Survived'] == 1).sum()
```

```python
#Task 2: Filter to show only female passengers who survived. How many are there?
mask = (df['Sex'] == 'female') & (df['Survived'] == 1)
mask.sum() 
```

```python
"""Task 3: Filter to show passengers who are either in first class or paid a fare greater than 100. 
Display the Name, Pclass, and Fare columns."""

mask = (df['Pclass'] == 1) | (df['Fare'] > 100) 
elite_persons = df[mask]
elite_persons[['Name', 'Pclass', 'Fare']].head(10)
```

```python
"""Task 4: Filter to show passengers who embarked at Southampton ('S') AND are older than 50. 
Display their Name, Age, and Embarked.
"""
mask = (df['Embarked'] == 'S') & (df['Age'] > 50)
older_southampton = df[mask]
older_southampton[['Name', 'Age', 'Embarked']].head()
```

```python
"""Task 5: Use .loc with a Boolean mask to select the Name and Age of all passengers who have a missing Cabin value. 
(Hint: df['Cabin'].isna())"""

missing_cabin = df.loc[df['Cabin'].isna(), ['Name', 'Age']]
missing_cabin.head()
```

```python
"""Task 6 (Bonus): Find all passengers whose name contains 'Master' (a title often given to young boys). 
Display their Name and Age."""

master_mask = df['Name'].str.contains('Master')
master_passenger = df[master_mask]
master_passenger[['Name', 'Age']].head()
```
