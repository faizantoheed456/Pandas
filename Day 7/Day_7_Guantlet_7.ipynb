```python
#Gauntlet #7: The Reshaping Challenge
import pandas as pd
df = pd.read_csv('titanic.csv')
```

```python
# Task 1: Melt to Long Format
# Create wide summary: average Fare by Pclass and Sex
wide_summary = df.groupby(['Pclass', 'Sex'])['Fare'].mean().unstack().reset_index()
print("Task 1 - Wide summary (before melt):")
print(wide_summary)

# Melt back to long format
long_summary = wide_summary.melt(id_vars='Pclass', var_name='Sex', value_name='Avg_Fare')
print("\nTask 1 - Melted long format:")
print(long_summary)
```

```python
# Task 2: Pivot Without Aggregation
pivoted_back = long_summary.pivot(index='Pclass', columns='Sex', values='Avg_Fare')
print("\nTask 2 - Pivoted back (no aggregation):")
print(pivoted_back)
```

```python
# Task 3: Pivot Table with Margins
pivot_median = pd.pivot_table(df, values='Age', index='Pclass', columns='Embarked',
                              aggfunc='median', margins=True, margins_name='Total',
                              fill_value=0)
print("\nTask 3 - Pivot table with margins:")
print(pivot_median)
```

```python
# Task 4: Stack and Unstack
pivot_multi = pd.pivot_table(df, values='Survived', index='Pclass',
                             columns=['Sex', 'Embarked'], aggfunc='mean')
print("\nTask 4 - Original pivot (MultiIndex columns):")
print(pivot_multi)

# Stack the 'Sex' level from columns to rows
stacked = pivot_multi.stack(level='Sex')
print("\nTask 4 - After stacking Sex level:")
print(stacked)
```

```python
# Task 5: Wide to Long
# Simulate wide DataFrame
wide = pd.pivot_table(df, values='Age', index='PassengerId',
                      columns='Pclass', aggfunc='mean').reset_index()
wide.columns = ['PassengerId', 'Class1', 'Class2', 'Class3']
print("\nTask 5 - Simulated wide DataFrame (first 5 rows):")
print(wide.head())

# Use pd.wide_to_long()
long_wide = pd.wide_to_long(wide, stubnames='Class', i='PassengerId',
                            j='Pclass', sep='', suffix='\d+').reset_index()
# Rename 'Class' to 'Age' for clarity
long_wide.rename(columns={'Class': 'Age'}, inplace=True)
print("\nTask 5 - After wide_to_long():")
print(long_wide.head())
```

```python
# Task 6 (Bonus): Pivot back to wide
# Create a column combining 'Class' and 'Pclass' to reconstruct original column names
long_wide['col'] = 'Class' + long_wide['Pclass'].astype(str)
wide_recreated = long_wide.pivot(index='PassengerId', columns='col', values='Age').reset_index()
wide_recreated.columns.name = None  # Remove columns name
print("\nTask 6 (Bonus) - Recreated wide format:")
print(wide_recreated.head())
```
