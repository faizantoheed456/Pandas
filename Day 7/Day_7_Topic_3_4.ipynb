```python
#Day 7, Topic 3: .pivot_table()
import pandas as pd
```

```python
#Basic .pivot_table()
df = pd.read_csv('titanic.csv')
# Average Fare by Pclass (rows) and Embarked (columns)
pivot = pd.pivot_table(df, values='Fare', index='Pclass', columns='Embarked', aggfunc='mean')
pivot
```

```python
#Multiple Aggregation Functions
pivot_mulit = pd.pivot_table(df, values='Fare', index='Pclass', columns='Embarked', aggfunc=['mean', 'median', 'count'])
pivot_mulit
```

```python
#Multiple values Columns
pivot_multi_col = pd.pivot_table(df, values=['Fare', 'Age'], index='Pclass', columns='Sex', aggfunc={'Fare': 'mean', 'Age': 'median'})
pivot_multi_col
```

```python
#Margins (Totals)
pivot_margins = pd.pivot_table(df, values='Survived', index='Pclass', columns='Sex', aggfunc='mean', margins=True, margins_name='Total')
pivot_margins
```

```python
#Filling Missing Values
pivot_filled = pd.pivot_table(df, values='Fare', index='Pclass', columns='Embarked', aggfunc='sum', fill_value=0)
pivot_filled
```

```python
#Grouping by Multiple Index/Columns
pivot_hier = pd.pivot_table(df, values='Fare', index=['Pclass', 'Sex'], columns='Embarked', aggfunc='mean')
pivot_hier
```

```python
#Practice Activity: .pivot_table() on Titanic
import pandas as pd
df = pd.read_csv('titanic.csv')
```

```python
"""Task 1: Create a pivot table showing the average Age for each combination of Pclass 
(rows) and Survived (columns). Use aggfunc='mean'."""

pivot_age = pd.pivot_table(df, index='Pclass', columns='Survived', values='Age', aggfunc='mean')
pivot_age
```

```python
#Day 7, Topic 4: .stack() and .unstack()
import pandas as pd
```

```python
#Creating a MultiIndex DataFrame to Work With
data = {
    'Region': ['North', 'North', 'South', 'South'],
    'Year': [2022, 2023, 2022, 2023],
    'Sales': [100, 120, 150, 170]
}
df = pd.DataFrame(data)

# Create a MultiIndex on Region and Year
df_multi = df.set_index(['Region', 'Year'])
print("Original MultiIndex DataFrame:")
df_multi
```

```python
unstacked = df_multi.unstack()
unstacked
```

```python
unstacked_region = df_multi.unstack(level=0)
unstacked_region
```

```python
#.stack() – Move Column Index to Rows
stacked_back = unstacked.stack()
stacked_back
```

```python
#Stacking/Unstacking with Multiple Levels
# Create a DataFrame with MultiIndex columns
cols = pd.MultiIndex.from_product([['Q1', 'Q2'], ['Units', 'Revenue']])
df_col_multi = pd.DataFrame([[10, 100, 12, 120],
                              [15, 150, 18, 180]],
                            index=['Product A', 'Product B'],
                            columns=cols)
print("DataFrame with MultiIndex columns:")
df_col_multi
```

```python
stacked_cols = df_col_multi.stack(level=1)
stacked_cols
```

```python
#Practice Activity: .stack() and .unstack()
import pandas as pd
df = pd.read_csv('titanic.csv')

# Create a pivot table with MultiIndex rows
pivot = pd.pivot_table(df, values='Age', index=['Pclass', 'Sex'], columns='Survived', aggfunc='mean')
pivot
```

```python
#Task 1: Use .unstack() on pivot to move the Survived level from columns to rows. Observe the resulting structure.

unstacked = pivot.unstack()
unstacked
```

```python
#Task 2: Take the result from Task 1 and use .stack() to return it to its original form.

stacked = unstacked.stack()
stacked
```

```python
"""Task 3: Create a pivot table with index='Embarked' and columns=['Pclass', 'Sex'] 
(values='Fare', aggfunc='median'). Then use .stack(level='Sex') to move the Sex level into the rows."""

pivot = pd.pivot_table(df, index='Embarked', columns=['Pclass', 'Sex'], values='Fare', aggfunc='median')
stacked = pivot.stack(level='Sex')
stacked
```

```python
"""Task 4 (Bonus): Using the pivot table from Task 3 (before stacking), use .unstack() on the 
row index (Embarked) to move it into the columns. What does the resulting DataFrame look like?"""

unstacked = stacked.unstack(level='Embarked')
unstacked
```
