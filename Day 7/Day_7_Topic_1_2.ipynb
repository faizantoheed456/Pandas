```python
#Day 7, Topic 1: .melt() – Unpivoting Wide Data to Long Format

import pandas as pd
```

```python
#Basic .melt()
wide_df = pd.DataFrame({
    'Salesperson': ['Faizan', 'Faisal', 'Usman'],
    'Q1': [100, 150, 200],
    'Q2': [110, 160, 210],
    'Q3': [120, 170, 220],
    'Q4': [130, 180, 230]
})

wide_df
```

```python
long_df = wide_df.melt(
    id_vars=['Salesperson'],
    var_name='Quarter',
    value_name='Sales'
)

long_df
```

```python
# Melt only Q1 and Q2, ignore Q3 and Q4
melted_partial = wide_df.melt(id_vars=['Salesperson'], value_vars=['Q1', 'Q2'], var_name='Quarter', value_name='Sales')

melted_partial
```

```python
#Real‑World Use Case – Titanic
# Simulate a wide version: aggregate survival counts by class and sex

df = pd.read_csv('titanic.csv')

# Create a pivot table (wide format)
pivot = pd.pivot_table(df, values='PassengerId', index='Pclass', columns='Sex', aggfunc='count')
pivot.columns = ['Female', 'Male']  # simplify column names
pivot = pivot.reset_index()
pivot.columns.name = None
print("Wide pivot table:")
pivot
```

```python
long_pivot = pivot.melt(id_vars=['Pclass'], var_name='Sex', value_name='Count')
long_pivot
```

```python
#Practice: .melt() on Titanic Data
import pandas as pd

df = pd.read_csv('titanic.csv')

# Create a wide summary: average Fare per Pclass and Embarked
wide = pd.pivot_table(df, values='Fare', index='Pclass', columns='Embarked', aggfunc='mean', fill_value=0)
wide = wide.reset_index()
wide.columns.name = None
wide
```

```python
"""Melt wide back to long format. Keep 'Pclass' as the identifier. 
The column headers ('C', 'Q', 'S') should become a new column named 'Port', 
and the fare values should be named 'Avg_Fare'. Display the result."""

long_pivot = wide.melt(id_vars=['Pclass'], value_vars=['C', 'Q', 'S'], var_name='Port', value_name='Avg_Fare')

long_pivot
```

```python
#Day 7, Topic 2: .pivot() – Reshaping Long Data to Wide Format
import pandas as pd
```

```python
#Basic .pivot() Syntax
long_df = pd.DataFrame({
    'Salesperson': ['Alice', 'Alice', 'Alice', 'Alice',
                    'Bob', 'Bob', 'Bob', 'Bob'],
    'Quarter': ['Q1', 'Q2', 'Q3', 'Q4',
                'Q1', 'Q2', 'Q3', 'Q4'],
    'Sales': [100, 110, 120, 130,
              150, 160, 170, 180]
})
print("Long format:")
long_df
```

```python
wide_df = long_df.pivot(index='Salesperson', columns='Quarter', values='Sales')
wide_df
```

```python
#Practice: .pivot() on Titanic Data
import pandas as pd

df = pd.read_csv('titanic.csv')

# Long summary: count of passengers by Pclass, Sex, and Survived
long_counts = df.groupby(['Pclass', 'Sex', 'Survived'])['PassengerId'].count().reset_index()
long_counts.columns = ['Pclass', 'Sex', 'Survived', 'Count']
long_counts.head()
```

```python
"""Task:
Pivot long_counts so that:
Rows are Pclass
Columns are Sex
Values are Count
Display the result."""

# Pivot the summary data
# Rows become Pclass, Columns become Sex, and the Values are the Counts
wide_df = pd.pivot_table(
    long_counts, 
    index='Pclass', 
    columns='Sex', 
    values='Count', 
    aggfunc='sum'
)

# Display the result
wide_df
```
